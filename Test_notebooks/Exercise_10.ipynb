# Problem 1: Solve the system of linear equations

Solve the following system of linear equations:
$$
\begin{aligned}
x + 2y + 3z &= 5 \\
2y + 3z &= 4 \\
3z &= 3
\end{aligned}
$$

### Step 1: Write the system in matrix form
The system can be written as:
$$
\begin{pmatrix}
1 & 2 & 3 \\
0 & 2 & 3 \\
0 & 0 & 3
\end{pmatrix}
\begin{pmatrix}
x \\
y \\
z
\end{pmatrix}
=
\begin{pmatrix}
5 \\
4 \\
3
\end{pmatrix}
$$
A = \begin{pmatrix} 1 & 2 & 3 \\ 0 & 2 & 3 \\ 0 & 0 & 3 \end{pmatrix}
B = \begin{pmatrix} 5 \\ 4 \\ 3 \end{pmatrix}

#### Step 2: Find the inverse of matrix \( A \)
Matrix \( A \) is triangular, so the inverse is easy to find. For a triangular matrix, the inverse can be computed by inverting the diagonal elements. The inverse of \( A \) is:
$$
A^{-1} =
\begin{pmatrix}
1 & 0 & 0 \\
0 & \frac{1}{2} & -\frac{1}{6} \\
0 & 0 & \frac{1}{3}
\end{pmatrix}
$$

#### Step 3: Multiply \( A^{-1} \) by \( b \)
Now, multiply \( A^{-1} \) by \( b \):
$$
A^{-1} \cdot b =
\begin{pmatrix}
1 & 0 & 0 \\
0 & \frac{1}{2} & -\frac{1}{6} \\
0 & 0 & \frac{1}{3}
\end{pmatrix}
\begin{pmatrix}
5 \\
4 \\
3
\end{pmatrix}
=
\begin{pmatrix}
5 \\
\frac{1}{2} \cdot 4 - \frac{1}{6} \cdot 3 \\
\frac{1}{3} \cdot 3
\end{pmatrix}
=
\begin{pmatrix}
5 \\
2 \\
1
\end{pmatrix}
$$
Thus, the solution is:
$$
x = 5, \quad y = 2, \quad z = 1
$$

# Problem 2: Solve the system of linear equations
###Solve the following system of linear equations:
$$
\begin{aligned}
x_1 + 2x_2 + 3x_3 &= 41 \\
4x_1 + 5x_2 + 6x_3 &= 93 \\
7x_1 + 8x_2 + 9x_3 &= 145
\end{aligned}
$$

####Step 1: Write the system in matrix form
The system can be written as:
$$
\begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2 \\
x_3
\end{pmatrix}
=
\begin{pmatrix}
41 \\
93 \\
145
\end{pmatrix}
$$
A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{pmatrix}
B = \begin{pmatrix} 41 \\ 93 \\ 145 \end{pmatrix}

#### Step 2: Find the determinant of matrix \( A \)
We first check if matrix \( A \) is invertible by computing its determinant. If the determinant is zero, the matrix is not invertible, and we cannot proceed.

The determinant of \( A \) is:
$$
\text{det}(A) = 1 \cdot \left| \begin{matrix} 5 & 6 \\ 8 & 9 \end{matrix} \right| - 2 \cdot \left| \begin{matrix} 4 & 6 \\ 7 & 9 \end{matrix} \right| + 3 \cdot \left| \begin{matrix} 4 & 5 \\ 7 & 8 \end{matrix} \right|
$$
Now compute each of the 2x2 determinants:
$$
\left| \begin{matrix} 5 & 6 \\ 8 & 9 \end{matrix} \right| = 5 \cdot 9 - 6 \cdot 8 = 45 - 48 = -3
$$
$$
\left| \begin{matrix} 4 & 6 \\ 7 & 9 \end{matrix} \right| = 4 \cdot 9 - 6 \cdot 7 = 36 - 42 = -6
$$
$$
\left| \begin{matrix} 4 & 5 \\ 7 & 8 \end{matrix} \right| = 4 \cdot 8 - 5 \cdot 7 = 32 - 35 = -3
$$
Now substitute these values back into the determinant formula:
$$
\text{det}(A) = 1 \cdot (-3) - 2 \cdot (-6) + 3 \cdot (-3) = -3 + 12 - 9 = 0
$$
Since the determinant of matrix \( A \) is zero, the matrix is singular and does not have an inverse. Therefore, this system cannot be solved using the inverse matrix method.
