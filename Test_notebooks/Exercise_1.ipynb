### Basic Operations on Matrices

#### Matrix Addition: \( A + B \)

Matrix \( A \):
$$
A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}
$$

Matrix \( B \):
$$
B = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}
$$

The addition \( A + B \) is:
$$
A + B = \begin{pmatrix} 1 + 5 & 2 + 6 \\ 3 + 7 & 4 + 8 \end{pmatrix} = \begin{pmatrix} 6 & 8 \\ 10 & 12 \end{pmatrix}
$$

#### Matrix Subtraction: \( B - A \)

The subtraction \( B - A \) is:
$$
B - A = \begin{pmatrix} 5 - 1 & 6 - 2 \\ 7 - 3 & 8 - 4 \end{pmatrix} = \begin{pmatrix} 4 & 4 \\ 4 & 4 \end{pmatrix}
$$

#### Matrix Addition: \( A + C \)

Matrix \( C \):
$$
C = \begin{pmatrix} -1 & 2 \\ 3 & 0 \end{pmatrix}
$$

The addition \( A + C \) is:
$$
A + C = \begin{pmatrix} 1 + (-1) & 2 + 2 \\ 3 + 3 & 4 + 0 \end{pmatrix} = \begin{pmatrix} 0 & 4 \\ 6 & 4 \end{pmatrix}
$$

#### Matrix Addition: \( D + E \)

Matrix \( D \):
$$
D = \begin{pmatrix} -1 & 4 \\ 2 & 0 \\ 3 & 6 \end{pmatrix}
$$

Matrix \( E \):
$$
E = \begin{pmatrix} 1 & 2 \\ 4 & 5 \\ 7 & 8 \end{pmatrix}
$$

Since matrix \( D \) is a \( 3 \times 2 \) matrix and matrix \( E \) is also a \( 3 \times 2 \) matrix, their addition is valid. The addition \( D + E \) is:
$$
D + E = \begin{pmatrix} -1 + 1 & 4 + 2 \\ 2 + 4 & 0 + 5 \\ 3 + 7 & 6 + 8 \end{pmatrix} = \begin{pmatrix} 0 & 6 \\ 6 & 5 \\ 10 & 14 \end{pmatrix}
$$

##### Scalar Multiplication:

The scalar multiplication is:
$$
\frac{1}{2}A = \frac{1}{2} \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} = \begin{pmatrix} \frac{1}{2} & 1 \\ \frac{3}{2} & 2 \end{pmatrix}
$$

##### Scalar Multiplication: \( 2B \)

The scalar multiplication \( 2B \) is:
$$
2B = 2 \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix} = \begin{pmatrix} 10 & 12 \\ 14 & 16 \end{pmatrix}
$$

#### Scalar Multiplication: \( -3C \)

The scalar multiplication \( -3C \) is:
$$
-3C = -3 \begin{pmatrix} -1 & 2 \\ 3 & 0 \end{pmatrix} = \begin{pmatrix} 3 & -6 \\ -9 & 0 \end{pmatrix}
$$

Scalar Multiplication: \( 4D \)

#### The scalar multiplication \( 4D \) is:
$$
4D = 4 \begin{pmatrix} -1 & 4 \\ 2 & 0 \\ 3 & 6 \end{pmatrix} = \begin{pmatrix} -4 & 16 \\ 8 & 0 \\ 12 & 24 \end{pmatrix}
$$

#### Matrix Product: \( A \cdot B \)

The product \( A \cdot B \) is:
$$
A \cdot B = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} \cdot \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}
$$
Calculating element by element:
$$
A \cdot B = \begin{pmatrix} (1 \cdot 5 + 2 \cdot 7) & (1 \cdot 6 + 2 \cdot 8) \\ (3 \cdot 5 + 4 \cdot 7) & (3 \cdot 6 + 4 \cdot 8) \end{pmatrix} = \begin{pmatrix} 5 + 14 & 6 + 16 \\ 15 + 28 & 18 + 32 \end{pmatrix}
$$
$$
A \cdot B = \begin{pmatrix} 19 & 22 \\ 43 & 50 \end{pmatrix}
$$

#### Matrix Product: \( B \cdot A \)

The product \( B \cdot A \) is:
$$
B \cdot A = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix} \cdot \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}
$$
#### Calculating element by element:
$$
B \cdot A = \begin{pmatrix} (5 \cdot 1 + 6 \cdot 3) & (5 \cdot 2 + 6 \cdot 4) \\ (7 \cdot 1 + 8 \cdot 3) & (7 \cdot 2 + 8 \cdot 4) \end{pmatrix} = \begin{pmatrix} 5 + 18 & 10 + 24 \\ 7 + 24 & 14 + 32 \end{pmatrix}
$$
$$
B \cdot A = \begin{pmatrix} 23 & 34 \\ 31 & 46 \end{pmatrix}
$$

#### Matrix Product: \( A \cdot D \)

The product \( A \cdot D \) is:
$$
A \cdot D = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} \cdot \begin{pmatrix} -1 & 4 \\ 2 & 0 \\ 3 & 6 \end{pmatrix}
$$
Calculating element by element:
$$
A \cdot D = \begin{pmatrix} (1 \cdot -1 + 2 \cdot 4) & (1 \cdot 2 + 2 \cdot 0) \\ (3 \cdot -1 + 4 \cdot 4) & (3 \cdot 2 + 4 \cdot 0) \\ (1 \cdot 3 + 2 \cdot 6) & (3 \cdot 3 + 4 \cdot 6) \end{pmatrix}
$$
$$
A \cdot D = \begin{pmatrix} -1 + 8 & 2 + 0 \\ -3 + 16 & 6 + 0 \\ 3 + 12 & 9 + 24 \end{pmatrix}
$$
$$
A \cdot D = \begin{pmatrix} 7 & 2 \\ 13 & 6 \\ 15 & 33 \end{pmatrix}
$$

##### Matrix Product: \( D \cdot E \)

The product \( D \cdot E \) is:
$$
D \cdot E = \begin{pmatrix} -1 & 4 \\ 2 & 0 \\ 3 & 6 \end{pmatrix} \cdot \begin{pmatrix} 1 & 2 \\ 4 & 5 \\ 7 & 8 \end{pmatrix}
$$
Calculating element by element:
$$
D \cdot E = \begin{pmatrix} (-1 \cdot 1 + 4 \cdot 4 + 3 \cdot 7) & (-1 \cdot 2 + 4 \cdot 5 + 3 \cdot 8) \\ (4 \cdot 1 + 0 \cdot 4 + 6 \cdot 7) & (4 \cdot 2 + 0 \cdot 5 + 6 \cdot 8) \end{pmatrix}
$$
$$
D \cdot E = \begin{pmatrix} -1 + 16 + 21 & -2 + 20 + 24 \\ 4 + 0 + 42 & 8 + 0 + 48 \end{pmatrix}
$$
$$
D \cdot E = \begin{pmatrix} 28 & 32 \\ 46 & 56 \end{pmatrix}
$$
