## 4. Determinants from the Gauss Method and Triangular Matrices

Perform row and column operations to reduce the following matrices to an upper triangular form and calculate their determinants by taking the product of the diagonal elements.

$$
\mathbf{A} = \begin{pmatrix}
12 & 3 \\
-18 & -4
\end{pmatrix}\qquad\qquad
\mathbf{B} = \begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{pmatrix}
$$

# 1. Determinant of \( A \) using Gaussian Elimination

Let \( A ) = \begin{pmatrix} 12 & 3 \\ -18 & -4 \end{pmatrix} \

### Step 1: Row operations to achieve upper triangular form

We want to eliminate the element in the lower left corner. We can perform the following row operation:

$$
R_2 \rightarrow R_2 + \frac{3}{2} R_1
$$

This operation is intended to make the element in the second row, first column zero. Applying this:

$$
R_2 = \begin{pmatrix} -18 & -4 \end{pmatrix} + \frac{3}{2} \begin{pmatrix} 12 & 3 \end{pmatrix} = \begin{pmatrix} -18 & -4 \end{pmatrix} + \begin{pmatrix} 18 & 4.5 \end{pmatrix} = \begin{pmatrix} 0 & 0.5 \end{pmatrix}
$$

Thus, the matrix \( A \) becomes:

$$
A = \begin{pmatrix} 12 & 3 \\ 0 & 0.5 \end{pmatrix}
$$

### Step 2: Calculate the determinant

The determinant of an upper triangular matrix is the product of the diagonal elements:

$$
\text{det}(A) = 12 \cdot 0.5 = 6
$$

Thus, the determinant of \( A \) is:

$$
\boxed{6}
$$

# 2. Determinant of \( B \) using Gaussian Elimination

Let \( B ) = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{pmatrix} \

### Step 1: Row operations to achieve upper triangular form

We will use row operations to eliminate the elements below the diagonal.

First, eliminate the element in the second row, first column by performing the row operation:

$$
R_2 \rightarrow R_2 - 4R_1
$$

This gives:

$$
R_2 = \begin{pmatrix} 4 & 5 & 6 \end{pmatrix} - 4 \cdot \begin{pmatrix} 1 & 2 & 3 \end{pmatrix} = \begin{pmatrix} 4 & 5 & 6 \end{pmatrix} - \begin{pmatrix} 4 & 8 & 12 \end{pmatrix} = \begin{pmatrix} 0 & -3 & -6 \end{pmatrix}
$$

Next, eliminate the element in the third row, first column by performing the row operation:

$$
R_3 \rightarrow R_3 - 7R_1
$$

This gives:

$$
R_3 = \begin{pmatrix} 7 & 8 & 9 \end{pmatrix} - 7 \cdot \begin{pmatrix} 1 & 2 & 3 \end{pmatrix} = \begin{pmatrix} 7 & 8 & 9 \end{pmatrix} - \begin{pmatrix} 7 & 14 & 21 \end{pmatrix} = \begin{pmatrix} 0 & -6 & -12 \end{pmatrix}
$$

Now, the matrix \( B \) looks like this:

$$
B = \begin{pmatrix} 1 & 2 & 3 \\ 0 & -3 & -6 \\ 0 & -6 & -12 \end{pmatrix}
$$

### Step 2: Eliminate the element in the third row, second column

We now eliminate the element in the third row, second column. Perform the row operation:

$$
R_3 \rightarrow R_3 - 2R_2
$$

This gives:

$$
R_3 = \begin{pmatrix} 0 & -6 & -12 \end{pmatrix} - 2 \cdot \begin{pmatrix} 0 & -3 & -6 \end{pmatrix} = \begin{pmatrix} 0 & -6 & -12 \end{pmatrix} - \begin{pmatrix} 0 & -6 & -12 \end{pmatrix} = \begin{pmatrix} 0 & 0 & 0 \end{pmatrix}
$$

Thus, the matrix \( B \) becomes:

$$
B = \begin{pmatrix} 1 & 2 & 3 \\ 0 & -3 & -6 \\ 0 & 0 & 0 \end{pmatrix}
$$

### Step 3: Calculate the determinant

The determinant of an upper triangular matrix is the product of the diagonal elements:

$$
\text{det}(B) = 1 \cdot (-3) \cdot 0 = 0
$$

Thus, the determinant of \( B \) is:

$$
\boxed{0}
$$
