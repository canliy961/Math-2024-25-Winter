### 1. Divide the line segment connecting points \(A(-1, 2)\) and \(B(3, -2)\) in the ratio \(1:3\). Illustrate the result on a graph.

The coordinates of the point \(P\) dividing the line segment \(AB\) in the ratio \(1:3\) can be found using the section formula. The section formula for dividing a line segment \(AB\) in the ratio \(m:n\) is given by:

$$
P = \left( \frac{n \cdot x_1 + m \cdot x_2}{m+n}, \frac{n \cdot y_1 + m \cdot y_2}{m+n} \right)
$$

For points \(A(-1, 2)\) and \(B(3, -2)\) and the ratio \(1:3\), we have:

$$
P = \left( \frac{3 \cdot (-1) + 1 \cdot 3}{1+3}, \frac{3 \cdot 2 + 1 \cdot (-2)}{1+3} \right)
$$

$$
P = \left( \frac{-3 + 3}{4}, \frac{6 - 2}{4} \right)
$$

$$
P = \left( \frac{0}{4}, \frac{4}{4} \right) = (0, 1)
$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the coordinates of points A, B, and P
A = np.array([-1, 2])
B = np.array([3, -2])
P = np.array([0, 1])

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the points A, B, and P
ax.plot(A[0], A[1], 'ro', label='A(-1, 2)')
ax.plot(B[0], B[1], 'bo', label='B(3, -2)')
ax.plot(P[0], P[1], 'go', label='P(0, 1)')

# Plot the line segment AB
ax.plot([A[0], B[0]], [A[1], B[1]], 'k--', label='Line segment AB')

# Plot the line segment AP and BP
ax.plot([A[0], P[0]], [A[1], P[1]], 'g--', label='AP (1:3 ratio)')
ax.plot([B[0], P[0]], [B[1], P[1]], 'g--', label='BP (1:3 ratio)')

# Add labels for the points
ax.text(A[0], A[1], ' A(-1, 2)', fontsize=12, verticalalignment='bottom', horizontalalignment='right')
ax.text(B[0], B[1], ' B(3, -2)', fontsize=12, verticalalignment='top', horizontalalignment='left')
ax.text(P[0], P[1], ' P(0, 1)', fontsize=12, verticalalignment='top', horizontalalignment='center')

# Set the limits of the graph
ax.set_xlim(-2, 4)
ax.set_ylim(-3, 3)

# Set the aspect of the plot to be equal
ax.set_aspect('equal')

# Add grid
ax.grid(True)

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

### 2. Project vector \(\mathbf{a} = (3, 4)\) onto the \(OX\) and \(OY\) axes. Illustrate the result on a graph.

The projection of a vector \(\mathbf{a} = (3, 4)\) onto the \(OX\)-axis (the \(x\)-axis) is simply the vector with the same \(x\)-component and a \(y\)-component of zero. This is done by setting the \(y\)-component of the vector to zero:

$$
\text{Proj}_{OX} \mathbf{a} = (3, 0)
$$

Similarly, the projection of \(\mathbf{a} = (3, 4)\) onto the \(OY\)-axis (the \(y\)-axis) is the vector with the same \(y\)-component and an \(x\)-component of zero. This is done by setting the \(x\)-component of the vector to zero:

$$
\text{Proj}_{OY} \mathbf{a} = (0, 4)
$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Vectors
a = np.array([3, 4])
proj_ox = np.array([3, 0])
proj_oy = np.array([0, 4])

# Plot
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\mathbf{a}$')
plt.quiver(0, 0, proj_ox[0], proj_ox[1], angles='xy', scale_units='xy', scale=1, color='g', label=r'$\text{Proj}_{OX} \mathbf{a}$')
plt.quiver(0, 0, proj_oy[0], proj_oy[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\text{Proj}_{OY} \mathbf{a}$')

# Labels and limits
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

### 3. Project vector \(\mathbf{a} = (2,3)\) onto vector \(\mathbf{b} = (1, 1)\). Illustrate the result on a graph.

The projection of vector \(\mathbf{a} = (2, 3)\) onto vector \(\mathbf{b} = (1, 1)\) is given by the formula:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{\mathbf{b} \cdot \mathbf{b}} \mathbf{b}
$$

First, calculate the dot products:

$$
\mathbf{a} \cdot \mathbf{b} = (2)(1) + (3)(1) = 2 + 3 = 5
$$

$$
\mathbf{b} \cdot \mathbf{b} = (1)(1) + (1)(1) = 1 + 1 = 2
$$

Now, compute the projection:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{5}{2} \mathbf{b} = \frac{5}{2} (1, 1) = \left( \frac{5}{2}, \frac{5}{2} \right)
$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the vectors
a = np.array([2, 3])
b = np.array([1, 1])

# Calculate the projection of a onto b
projection = (np.dot(a, b) / np.dot(b, b)) * b

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the vectors a, b, and the projection
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a (2, 3)')
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector b (1, 1)')
ax.quiver(0, 0, projection[0], projection[1], angles='xy', scale_units='xy', scale=1, color='g', label='Proj_b a')

# Set the limits of the graph
ax.set_xlim(0, 3)
ax.set_ylim(0, 4)

# Set the aspect of the plot to be equal
ax.set_aspect('equal')

# Add grid
ax.grid(True)

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

### 4. Project vector \(\mathbf{b} = (1, 1)\) onto vector \(\mathbf{a} = (2, 3)\). Illustrate the result on a graph.

The projection of vector \(\mathbf{b} = (1, 1)\) onto vector \(\mathbf{a} = (2, 3)\) is given by:

$$
\text{Proj}_{\mathbf{a}} \mathbf{b} = \frac{\mathbf{b} \cdot \mathbf{a}}{\mathbf{a} \cdot \mathbf{a}} \mathbf{a}
$$

First, calculate the dot products:

$$
\mathbf{b} \cdot \mathbf{a} = (1)(2) + (1)(3) = 2 + 3 = 5
$$

$$
\mathbf{a} \cdot \mathbf{a} = (2)(2) + (3)(3) = 4 + 9 = 13
$$

Now, compute the projection:

$$
\text{Proj}_{\mathbf{a}} \mathbf{b} = \frac{5}{13} \mathbf{a} = \frac{5}{13} (2, 3) = \left( \frac{10}{13}, \frac{15}{13} \right)
$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the vectors
a = np.array([2, 3])
b = np.array([1, 1])

# Calculate the projection of b onto a
projection = (np.dot(b, a) / np.dot(a, a)) * a

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the vectors a, b, and the projection
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a (2, 3)')
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector b (1, 1)')
ax.quiver(0, 0, projection[0], projection[1], angles='xy', scale_units='xy', scale=1, color='g', label='Proj_a b')

# Set the limits of the graph
ax.set_xlim(0, 3)
ax.set_ylim(0, 4)

# Set the aspect of the plot to be equal
ax.set_aspect('equal')

# Add grid
ax.grid(True)

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
