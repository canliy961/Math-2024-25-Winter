## 7. Linear Equations old school

Solve the following systems of equations without using matrices:

* $3x-2y=5, \quad 2x+3y=7$,
* $2x-3y=10, \quad 4x+5y=20$,
* $2x - y + z = 3, \quad x + 2y - z = 1, \quad 3x - y + 2z = 11$.
* $2x-3y+4z+2t=2, \quad 3x+2y-5z+3t=3, \quad 4x-3y+2z-5t=4, \quad 5x+4y-3z+2t=5$.


### 1. Solve the system:
$$
3x - 2y = 5 \quad \text{(1)}
$$
$$
2x + 3y = 7 \quad \text{(2)}
$$

#### Step 1: Solve equation (1) for \( x \):
$$
3x = 5 + 2y \quad \Rightarrow \quad x = \frac{5 + 2y}{3}
$$

#### Step 2: Substitute this expression for \( x \) in equation (2):
Substitute \( x = \frac{5 + 2y}{3} \) into \( 2x + 3y = 7 \):
$$
2\left( \frac{5 + 2y}{3} \right) + 3y = 7
$$
Multiply through by 3 to eliminate the denominator:
$$
2(5 + 2y) + 9y = 21 \quad \Rightarrow \quad 10 + 4y + 9y = 21 \quad \Rightarrow \quad 10 + 13y = 21
$$
Simplify and solve for \( y \):
$$
13y = 11 \quad \Rightarrow \quad y = \frac{11}{13}
$$

#### Step 3: Substitute \( y = \frac{11}{13} \) back into \( x = \frac{5 + 2y}{3} \):
$$
x = \frac{5 + 2\left( \frac{11}{13} \right)}{3} = \frac{5 + \frac{22}{13}}{3} = \frac{\frac{65}{13} + \frac{22}{13}}{3} = \frac{87}{39} = \frac{29}{13}
$$

#### Final Answer for the first system:
$$
x = \frac{29}{13}, \quad y = \frac{11}{13}
$$

### 2. Solve the system:
$$
2x - 3y = 10 \quad \text{(3)}
$$
$$
4x + 5y = 20 \quad \text{(4)}
$$

#### Step 1: Solve equation (3) for \( x \):
$$
2x = 10 + 3y \quad \Rightarrow \quad x = \frac{10 + 3y}{2}
$$

#### Step 2: Substitute this expression for \( x \) in equation (4):
Substitute \( x = \frac{10 + 3y}{2} \) into \( 4x + 5y = 20 \):
$$
4\left( \frac{10 + 3y}{2} \right) + 5y = 20
$$
Multiply through by 2 to eliminate the fraction:
$$
2(10 + 3y) + 5y = 20 \quad \Rightarrow \quad 20 + 6y + 5y = 20 \quad \Rightarrow \quad 20 + 11y = 20
$$
Simplify and solve for \( y \):
$$
11y = 0 \quad \Rightarrow \quad y = 0
$$

#### Step 3: Substitute \( y = 0 \) back into \( x = \frac{10 + 3y}{2} \):
$$
x = \frac{10 + 3(0)}{2} = \frac{10}{2} = 5
$$

#### Final Answer for the second system:
$$
x = 5, \quad y = 0
$$

### 3. Solve the system:
$$
2x - y + z = 3 \quad \text{(5)}
$$
$$
x + 2y - z = 1 \quad \text{(6)}
$$
$$
3x - y + 2z = 11 \quad \text{(7)}
$$

#### Step 1: Solve equation (5) for \( x \):
$$
2x = 3 + y - z \quad \Rightarrow \quad x = \frac{3 + y - z}{2}
$$

#### Step 2: Substitute this expression for \( x \) in equations (6) and (7).

Substitute in equation (6):
$$
\left( \frac{3 + y - z}{2} \right) + 2y - z = 1
$$
Multiply through by 2 to eliminate the fraction:
$$
3 + y - z + 4y - 2z = 2 \quad \Rightarrow \quad 3 + 5y - 3z = 2 \quad \Rightarrow \quad 5y - 3z = -1 \quad \text{(8)}
$$

Substitute in equation (7):
$$
3\left( \frac{3 + y - z}{2} \right) - y + 2z = 11
$$
Multiply through by 2:
$$
9 + 3y - 3z - 2y + 4z = 22 \quad \Rightarrow \quad 9 + y + z = 22 \quad \Rightarrow \quad y + z = 13 \quad \text{(9)}
$$

#### Step 3: Solve the system of two equations (8) and (9):

From equation (9), solve for \( y \):
$$
y = 13 - z
$$
Substitute \( y = 13 - z \) into equation (8):
$$
5(13 - z) - 3z = -1 \quad \Rightarrow \quad 65 - 5z - 3z = -1 \quad \Rightarrow \quad 65 - 8z = -1 \quad \Rightarrow \quad -8z = -66 \quad \Rightarrow \quad z = \frac{66}{8} = \frac{33}{4}
$$

#### Step 4: Substitute \( z = \frac{33}{4} \) back into \( y = 13 - z \):
$$
y = 13 - \frac{33}{4} = \frac{52}{4} - \frac{33}{4} = \frac{19}{4}
$$

#### Step 5: Substitute \( y = \frac{19}{4} \) and \( z = \frac{33}{4} \) back into \( x = \frac{3 + y - z}{2} \):
$$
x = \frac{3 + \frac{19}{4} - \frac{33}{4}}{2} = \frac{3 - \frac{14}{4}}{2} = \frac{\frac{12}{4} - \frac{14}{4}}{2} = \frac{-\frac{2}{4}}{2} = \frac{-2}{4} = \frac{-1}{4}
$$

#### Final Answer for the third system:
$$
x = \frac{-1}{4}, \quad y = \frac{19}{4}, \quad z = \frac{33}{4}
$$
