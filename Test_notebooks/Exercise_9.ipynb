## 9. Linear equations by Gauss Elimination

$$\begin{cases}
x + 2y - 2z = 4 \\
2x + y + z = 0 \\
3x + 2y + z = 1
\end{cases}
\quad
\begin{cases}
x + y + z - t = 2 \\
2x + y + z = 3 \\
-x + z - t = 0 \\
3x + 2y - z + 2t = -1
\end{cases}
\quad
\begin{cases}
x + y - z - t = 0 \\
2x + 3y - 2z + t = 4 \\
3x + 5z = 0 \\
-x + y - 3z + 2t = 3
\end{cases}
$$

### System 1

$$
\begin{aligned}
x + 2y - 2z &= 4 \\
2x + y + z &= 0 \\
3x + 2y + z &= 1
\end{aligned}
$$

Step 1: Write the augmented matrix.
$$
\left(\begin{array}{ccc|c}
1 & 2 & -2 & 4 \\
2 & 1 & 1 & 0 \\
3 & 2 & 1 & 1
\end{array}\right)
$$

Step 2: Perform row operations to create zeros below the leading entry in the first column.\\
Row 2: Subtract 2 times Row 1 from Row 2.\\
Row 3: Subtract 3 times Row 1 from Row 3.
$$
\left(\begin{array}{ccc|c}
1 & 2 & -2 & 4 \\
0 & -3 & 5 & -8 \\
0 & -4 & 7 & -11
\end{array}\right)
$$

Step 3: Create zeros below the leading entry in the second column.\\
Row 3: Subtract \(\frac{4}{3}\) times Row 2 from Row 3.
$$
\left(\begin{array}{ccc|c}
1 & 2 & -2 & 4 \\
0 & -3 & 5 & -8 \\
0 & 0 & 1 & 1
\end{array}\right)
$$

Step 4: Back substitution.\\
From Row 3: \( z = 1 \).\\
From Row 2: \( -3y + 5z = -8 \Rightarrow -3y + 5(1) = -8 \Rightarrow y = 3 \).\\
From Row 1: \( x + 2y - 2z = 4 \Rightarrow x + 2(3) - 2(1) = 4 \Rightarrow x = -2 \).

#### Final solution for System 1:

$$
x = -2, \quad y = 3, \quad z = 1
$$

### System 2

$$
\begin{aligned}
x + y + z - t &= 2 \\
2x + y + z &= 3 \\
-x + z - t &= 0 \\
3x + 2y - z + 2t &= -1
\end{aligned}
$$

Step 1: Write the augmented matrix.
$$
\left(\begin{array}{cccc|c}
1 & 1 & 1 & -1 & 2 \\
2 & 1 & 1 & 0 & 3 \\
-1 & 0 & 1 & -1 & 0 \\
3 & 2 & -1 & 2 & -1
\end{array}\right)
$$

Step 2: Perform row operations to create zeros below the leading entry in the first column.\\
Row 2: Subtract 2 times Row 1 from Row 2.\\
Row 3: Add Row 1 to Row 3.\\
Row 4: Subtract 3 times Row 1 from Row 4.
$$
\left(\begin{array}{cccc|c}
1 & 1 & 1 & -1 & 2 \\
0 & -1 & -1 & 2 & -1 \\
0 & 1 & 2 & -2 & 2 \\
0 & -1 & 2 & -1 & -7
\end{array}\right)
$$

Step 3: Create zeros below the leading entry in the second column.\\
Row 3: Add Row 2 to Row 3.\\
Row 4: Add Row 2 to Row 4.
$$
\left(\begin{array}{cccc|c}
1 & 1 & 1 & -1 & 2 \\
0 & -1 & -1 & 2 & -1 \\
0 & 0 & 1 & 0 & 1 \\
0 & 0 & 1 & 1 & -8
\end{array}\right)
$$

Step 4: Create a zero below the third column in Row 4.\\
Row 4: Add 5 times Row 3 to Row 4.
$$
\left(\begin{array}{cccc|c}
1 & 1 & 1 & -1 & 2 \\
0 & -1 & -1 & 2 & -1 \\
0 & 0 & 1 & 0 & 1 \\
0 & 0 & 0 & 7 & -3
\end{array}\right)
$$

Step 5: Back substitution.\\
From Row 4: \( 7t = -3 \Rightarrow t = -\frac{3}{7} \).\\
From Row 3: \( z = 1 \).\\
From Row 2: \( -y - z + 2t = -1 \Rightarrow -y - 1 + 2(-\frac{3}{7}) = -1 \Rightarrow y = \frac{10}{7} \).\\
From Row 1: \( x + y + z - t = 2 \Rightarrow x + \frac{10}{7} + 1 - (-\frac{3}{7}) = 2 \Rightarrow x = -\frac{6}{7} \).

#### Final solution for System 2:

$$
x = -\frac{6}{7}, \quad y = \frac{10}{7}, \quad z = 1, \quad t = -\frac{3}{7}
$$

### System 3

$$
\begin{aligned}
x + y - z - t &= 0 \\
2x + 3y - 2z + t &= 4 \\
3x + 5z &= 0 \\
-x + y - 3z + 2t &= 3
\end{aligned}
$$

Step 1: Write the augmented matrix.
$$
\left(\begin{array}{cccc|c}
1 & 1 & -1 & -1 & 0 \\
2 & 3 & -2 & 1 & 4 \\
3 & 0 & 5 & 0 & 0 \\
-1 & 1 & -3 & 2 & 3
\end{array}\right)
$$

Step 2: Perform row operations to create zeros below the leading entry in the first column.\\
Row 2: Subtract 2 times Row 1 from Row 2.\\
Row 3: Subtract 3 times Row 1 from Row 3.\\
Row 4: Add Row 1 to Row 4.
$$
\left(\begin{array}{cccc|c}
1 & 1 & -1 & -1 & 0 \\
0 & 1 & 0 & 3 & 4 \\
0 & -3 & 8 & 4 & 0 \\
0 & 2 & -4 & 1 & 3
\end{array}\right)
$$

Step 3: Create zeros below the second column.\\
Row 3: Add 3 times Row 2 to Row 3.\\
Row 4: Subtract 2 times Row 2 from Row 4.
$$
\left(\begin{array}{cccc|c}
1 & 1 & -1 & -1 & 0 \\
0 & 1 & 0 & 3 & 4 \\
0 & 0 & 8 & 12 & 12 \\
0 & 0 & -4 & -5 & -5
\end{array}\right)
$$

Step 4: Create a zero below the third column.\\
Row 4: Add \(\frac{1}{2}\) times Row 3 to Row 4.
$$
\left(\begin{array}{cccc|c}
1 & 1 & -1 & -1 & 0 \\
0 & 1 & 0 & 3 & 4 \\
0 & 0 & 8 & 12 & 12 \\
0 & 0 & 0 & 0 & 0
\end{array}\right)
$$

Step 5: Back substitution.\\
From Row 3: \( 8z + 12t = 12 \Rightarrow z + \frac{3}{2} t = \frac{3}{2} \Rightarrow z = \frac{3}{2} - \frac{3}{2} t \).\\
From Row 2: \( y + 3z = 4 \Rightarrow y + 3(\frac{3}{2} - \frac{3}{2} t) = 4 \Rightarrow y = \frac{7}{2} - \frac{9}{2} t \).\\
From Row 1: \( x + y - z - t = 0 \Rightarrow x + (\frac{7}{2} - \frac{9}{2} t) - (\frac{3}{2} - \frac{3}{2} t) - t = 0 \Rightarrow x = -\frac{1}{2} + \frac{5}{2} t \).

#### Final solution for System 3:

$$
x = -\frac{1}{2} + \frac{5}{2} t, \quad y = \frac{7}{2} - \frac{9}{2} t, \quad z = \frac{3}{2} - \frac{3}{2} t, \quad t \text{ is free}.
$$
