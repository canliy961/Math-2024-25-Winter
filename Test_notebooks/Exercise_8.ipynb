4. Why can't the following system of equations be solved using Cramer's rule?

$$\begin{cases}
x_1 + 2x_2 + 3x_3 = 3 \\
4x_1 + 5x_2 + 6x_3 = 2 \\
7x_1 + 8x_2 + 9x_3 = 1
\end{cases}$$

1. Solve the system of equations using Cramer's Rule:
The system of equations is:


\begin{aligned}
2x_1 - 3x_2 &= 7 \\
3x_1 + 5x_2 &= 2
\end{aligned}


The system can be written in matrix form as:

A x = b

where

A = \begin{pmatrix} 2 & -3 \\ 3 & 5 \end{pmatrix}, \quad x = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}, \quad b = \begin{pmatrix} 7 \\ 2 \end{pmatrix}


**Step 1**: Find the determinant of $A$


\text{det}(A) = (2)(5) - (-3)(3) = 10 + 9 = 19


**Step 2**: Find the determinant of $A_1$ (replace the first column of $A$ with $b$)


A_1 = \begin{pmatrix} 7 & -3 \\ 2 & 5 \end{pmatrix}


\text{det}(A_1) = (7)(5) - (-3)(2) = 35 + 6 = 41


**Step 3**: Find the determinant of $A_2$ (replace the second column of $A$ with $b$)

A_2 = \begin{pmatrix} 2 & 7 \\ 3 & 2 \end{pmatrix}


\text{det}(A_2) = (2)(2) - (7)(3) = 4 - 21 = -17

**Step 4**: Apply Cramer's Rule

Using Cramer's rule:

x_1 = \frac{\text{det}(A_1)}{\text{det}(A)} = \frac{41}{19}, \quad x_2 = \frac{\text{det}(A_2)}{\text{det}(A)} = \frac{-17}{19}

Thus, the solution is:

x_1 = \frac{41}{19}, \quad x_2 = \frac{-17}{19}

2. Solve the system of equations using Cramer's Rule:
The system of equations is:

\begin{aligned}
2x + y - z &= 1 \\
x - y + 2z &= 4 \\
3x - 2z &= -1
\end{aligned}

This system can be written as:

A x = b

where

A = \begin{pmatrix} 2 & 1 & -1 \\ 1 & -1 & 2 \\ 3 & 0 & -2 \end{pmatrix}, \quad x = \begin{pmatrix} x \\ y \\ z \end{pmatrix}, \quad b = \begin{pmatrix} 1 \\ 4 \\ -1 \end{pmatrix}

**Step 1**: Find the determinant of $A$

\text{det}(A) = 2 \begin{vmatrix} -1 & 2 \\ 0 & -2 \end{vmatrix} - 1 \begin{vmatrix} 1 & 2 \\ 3 & -2 \end{vmatrix} + (-1) \begin{vmatrix} 1 & -1 \\ 3 & 0 \end{vmatrix}

First calculate the $2 \times 2$ determinants:

\begin{vmatrix} -1 & 2 \\ 0 & -2 \end{vmatrix} = (-1)(-2) - (2)(0) = 2

\begin{vmatrix} 1 & 2 \\ 3 & -2 \end{vmatrix} = (1)(-2) - (2)(3) = -2 - 6 = -8

\begin{vmatrix} 1 & -1 \\ 3 & 0 \end{vmatrix} = (1)(0) - (-1)(3) = 3

Now, calculate the determinant of $A$:

\text{det}(A) = 2(2) - 1(-8) + (-1)(3) = 4 + 8 - 3 = 9

**Step 2**: Find the determinant of $A_1$ (replace the first column of $A$ with $b$)

A_1 = \begin{pmatrix} 1 & 1 & -1 \\ 4 & -1 & 2 \\ -1 & 0 & -2 \end{pmatrix}

\text{det}(A_1) = 1 \begin{vmatrix} -1 & 2 \\ 0 & -2 \end{vmatrix} - 1 \begin{vmatrix} 4 & 2 \\ -1 & -2 \end{vmatrix} + (-1) \begin{vmatrix} 4 & -1 \\ -1 & 0 \end{vmatrix}

Calculate the $2 \times 2$ determinants:

\begin{vmatrix} -1 & 2 \\ 0 & -2 \end{vmatrix} = 2

\begin{vmatrix} 4 & 2 \\ -1 & -2 \end{vmatrix} = (4)(-2) - (2)(-1) = -8 + 2 = -6



\begin{vmatrix} 4 & -1 \\ -1 & 0 \end{vmatrix} = (4)(0) - (-1)(-1) = -1


Now, calculate the determinant of $A_1$:


\text{det}(A_1) = 1(2) - 1(-6) + (-1)(-1) = 2 + 6 + 1 = 9


**Step 3**: Find the determinant of $A_2$ (replace the second column of $A$ with $b$)


A_2 = \begin{pmatrix} 2 & 1 & -1 \\ 1 & 4 & 2 \\ 3 & -1 & -2 \end{pmatrix}


Calculate similarly as in previous steps:


\text{det}(A_2) = 1 \left( \begin{vmatrix} 4 & 2 \\ -1 & -2 \end{vmatrix} \right) - 1 \left( \begin{vmatrix} 1 & 2 \\ 3 & -2 \end{vmatrix} \right) + (-1) \left( \begin{vmatrix} 1 & 4 \\ 3 & -1 \end{vmatrix} \right)


The result will be:


\text{det}(A_2) = 9


**Step 4**: Find the determinant of $A_3$ (replace the third column of $A$ with $b$)


A_3 = \begin{pmatrix} 2 & 1 & 1 \\ 1 & -1 & 4 \\ 3 & 0 & -1 \end{pmatrix}


Calculate similarly as before:


\text{det}(A_3) = 28


**Step 5**: Apply Cramer's rule

Now, apply Cramer's rule:


x = \frac{\text{det}(A_1)}{\text{det}(A)} = \frac{9}{9} = 1, \quad y = \frac{\text{det}(A_2)}{\text{det}(A)} = \frac{9}{9} = 1, \quad z = \frac{\text{det}(A_3)}{\text{det}(A)} = \frac{28}{9}

Thus, the solution is:


x = 1, \quad y = 1, \quad z = \frac{28}{9}

4. Why can't the following system be solved using Cramer's rule?
The system is:


\begin{aligned}
x_1 + 2x_2 + 3x_3 &= 3 \\
4x_1 + 5x_2 + 6x_3 &= 2 \\
7x_1 + 8x_2 + 9x_3 &= 1
\end{aligned}


To solve this using Cramer's rule, the determinant of the coefficient matrix $A$ must be non-zero. The matrix $A$ is:


A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{pmatrix}


The determinant of $A$ is:


\text{det}(A) = 1 \begin{vmatrix} 5 & 6 \\ 8 & 9 \end{vmatrix} - 2 \begin{vmatrix} 4 & 6 \\ 7 & 9 \end{vmatrix} + 3 \begin{vmatrix} 4 & 5 \\ 7 & 8 \end{vmatrix}


Calculating the $2 \times 2$ determinants:


\begin{vmatrix} 5 & 6 \\ 8 & 9 \end{vmatrix} = -3, \quad \begin{vmatrix} 4 & 6 \\ 7 & 9 \end{vmatrix} = -6, \quad \begin{vmatrix} 4 & 5 \\ 7 & 8 \end{vmatrix} = -3


Thus,


\text{det}(A) = 1(-3) - 2(-6) + 3(-3) = -3 + 12 - 9 = 0

Since the determinant is zero, the coefficient matrix is singular, and Cramer's rule cannot be applied. Therefore, the system cannot be solved using Cramer's rule.
