We are given the functions
$f(x) = e^x$
and
$g(x) = \ln(x)$,
and we need to compute
$f(g(x))$
and
$g(f(x))$.

### Step 1: Compute $f(g(x))$
We know that:

$f(x) = e^x$
and
$g(x) = \ln(x)$.

So, to compute
$f(g(x))$,
we substitute $g(x)$ into $f(x)$:

$$
f(g(x)) = f(\ln(x)) = e^{\ln(x)}
$$

Since
$e^{\ln(x)} = x$
(by the property of the natural logarithm and the exponential function being inverse functions), we have:

$$
f(g(x)) = x.
$$

### Step 2: Compute $g(f(x))$
Now, we compute
$g(f(x))$, which is
$g(e^x)$.
Substituting $e^x$ into $g(x)$:

$$
g(f(x)) = g(e^x) = \ln(e^x)
$$

Since
$\ln(e^x) = x$
(by the inverse property of the natural logarithm and the exponential function), we get:

$$
g(f(x)) = x.
$$

### Conclusion
We have:

$$
f(g(x)) = x
$$
and
$$
g(f(x)) = x.
$$

This means that both
$f(g(x)) = x$
and
$g(f(x)) = x$
result in $x$.
This is consistent with the fact that the functions
$f(x) = e^x$
and
$g(x) = \ln(x)$
are inverses of each other. Thus, applying one function after the other effectively "undoes" the operation, returning the original input $x$.

We are given the function
$$ f = \{ (1,7), (2,9), (3,11) \} $$
which represents a set of ordered pairs. To find the inverse function $f^{-1}$, we need to reverse the roles of the inputs and outputs in each pair.

### Step 1: Reverse the pairs
The inverse function $f^{-1}$ consists of the following steps:

- The pair $(1,7)$ in $f$ becomes $(7,1)$ in $f^{-1}$.
- The pair $(2,9)$ in $f$ becomes $(9,2)$ in $f^{-1}$.
- The pair $(3,11)$ in $f$ becomes $(11,3)$ in $f^{-1}$.

### Step 2: Write the inverse function
The inverse function $f^{-1}$ is:

$$ f^{-1} = \{ (7,1), (9,2), (11,3) \} $$

Thus, the inverse of the given function $f$ is:

$$ f^{-1} = \{ (7,1), (9,2), (11,3) \} $$


#5.
We are given the function
$$ f = \{ (1,7), (2,7), (3,11) \} $$
which represents a set of ordered pairs, where the first element in each pair represents the input (domain), and the second element represents the output (range). To find the inverse function $f^{-1}$, we need to reverse the roles of the inputs and outputs.

### Step 1: Reverse the pairs
The inverse function $f^{-1}$ is formed by swapping each input and output in the original pairs:

- The pair $(1,7)$ becomes $(7,1)$.
- The pair $(2,7)$ becomes $(7,2)$.
- The pair $(3,11)$ becomes $(11,3)$.

### Step 2: Check for uniqueness
In this case, notice that both $(1,7)$ and $(2,7)$ have the same output value of $7$. In an inverse function, each output must correspond to exactly one input. Since $7$ is mapped to both $1$ and $2$, the inverse function does not exist in the usual sense because it fails the "one-to-one" requirement.

### Conclusion
The function
$$ f = \{ (1,7), (2,7), (3,11) \} $$
is not one-to-one, so it does not have an inverse function $f^{-1}$.
