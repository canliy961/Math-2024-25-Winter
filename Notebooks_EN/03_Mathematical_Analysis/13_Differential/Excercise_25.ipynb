## 25. Differential Equations

### 1. Solve the following first-order ordinary differential equations:

#### (a) $ y'(x) = y(x) $

This is a separable differential equation:

$$ \frac{dy}{dx} = y $$

Rearrange:

$$ \frac{dy}{y} = dx $$

Integrate both sides:

$$ \ln |y| = x + C $$

Exponentiate:

$$ y = C e^x $$

where $ C $ is the integration constant.

#### (b) $ y'(x) = \frac{1}{2y(x)} $

Rearrange:

$$ 2y \frac{dy}{dx} = 1 $$

Integrate both sides:

$$ \int 2y dy = \int dx $$

$$ y^2 = x + C $$

Solve for $ y $:

$$ y = \pm \sqrt{x + C} $$

### 2. Solve the first-order ordinary differential equations using the method of separation of variables:

#### (a) $ \frac{dy}{dx} = \frac{x}{y} $

Rearrange:

$$ y dy = x dx $$

Integrate:

$$ \int y dy = \int x dx $$

$$ \frac{y^2}{2} = \frac{x^2}{2} + C $$

Solve for $ y $:

$$ y = \pm \sqrt{x^2 + 2C} $$

#### (b) $ \frac{dy}{dx} = \frac{y}{x} $

Rearrange:

$$ \frac{dy}{y} = \frac{dx}{x} $$

Integrate:

$$ \ln |y| = \ln |x| + C $$

Exponentiate:

$$ y = Cx $$

#### (c) $ \frac{dy}{dx} = xy $

Rearrange:

$$ \frac{dy}{y} = x dx $$

Integrate:

$$ \ln |y| = \frac{x^2}{2} + C $$

Exponentiate:

$$ y = C e^{x^2 / 2} $$

### 3. Solve the second-order ordinary differential equations:

#### (a) $ y''(x) + y'(x) = 0 $, with boundary conditions $ y(0) = 2 $, $ y'(0) = -1 $

The characteristic equation is:

$$ r^2 + r = 0 $$

Factorize:

$$ r(r + 1) = 0 $$

So, $ r = 0 $ or $ r = -1 $. The general solution is:

$$ y(x) = C_1 + C_2 e^{-x} $$

Using boundary conditions:

1. $ y(0) = 2 \Rightarrow C_1 + C_2 = 2 $
2. $ y'(x) = -C_2 e^{-x} $ gives $ y'(0) = -C_2 = -1 \Rightarrow C_2 = 1 $

Thus, $ C_1 = 1 $ and the solution is:

$$ y(x) = 1 + e^{-x} $$

#### (b) $ y''(x) - y(x) = 0 $, with boundary conditions $ y(0) = 2 $, $ y'(0) = 0 $

The characteristic equation is:

$$ r^2 - 1 = 0 $$

$$ (r - 1)(r + 1) = 0 $$

So, $ r = \pm 1 $. The general solution is:

$$ y(x) = C_1 e^x + C_2 e^{-x} $$

Using boundary conditions:

1. $ y(0) = 2 \Rightarrow C_1 + C_2 = 2 $
2. $ y'(x) = C_1 e^x - C_2 e^{-x} $, so $ y'(0) = C_1 - C_2 = 0 \Rightarrow C_1 = C_2 $

Solving $ C_1 + C_2 = 2 $ and $ C_1 = C_2 $, we get $ C_1 = C_2 = 1 $.

Thus, the solution is:

$$ y(x) = e^x + e^{-x} $$

#### (c) $ \frac{d^2 y(x)}{dx^2} = -\omega^2 y(x) $

The characteristic equation is:

$$ r^2 + \omega^2 = 0 $$

$$ r = \pm i\omega $$

The general solution is:

$$ y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x) $$

### 4. Check if $ \psi(t, x) = A \cos(\omega t + kx) $ is a solution of the wave equation:

The wave equation:

$$ \frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0 $$

Compute derivatives:

$$ \frac{\partial \psi}{\partial t} = -A \omega \sin(\omega t + kx) $$

$$ \frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx) $$

$$ \frac{\partial \psi}{\partial x} = -A k \sin(\omega t + kx) $$

$$ \frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx) $$

Substituting into the wave equation:

$$ -A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx)) = 0 $$

Factor out:

$$ -A \cos(\omega t + kx) (\omega^2 - v^2 k^2) = 0 $$

Since $ A \cos(\omega t + kx) \neq 0 $, we require:

$$ \omega^2 = v^2 k^2 $$

which holds since $ v = \frac{\omega}{k} $.

Thus, $ \psi(t, x) $ is a solution of the wave equation.
