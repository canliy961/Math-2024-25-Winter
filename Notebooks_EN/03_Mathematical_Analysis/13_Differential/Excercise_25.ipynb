## 1. Solve the following first-order ordinary differential equations:

   -    - $y'(x)= y(x)$

     The equation is separable. We can rewrite it as:
     $$
     \frac{dy}{dx} = y \quad \Rightarrow \quad \frac{dy}{y} = dx.
     $$
     Integrating both sides:
     $$
     \int \frac{1}{y} \, dy = \int 1 \, dx \quad \Rightarrow \quad \ln|y| = x + C.
     $$
     Exponentiating both sides:
     $$
     y(x) = Ce^x,
     $$
     where \( C \) is a constant determined by initial conditions.

   - $y'(x) = \frac{1}{2y(x)}$

     The equation is separable. We can rewrite it as:
     $$
     2y \, dy = dx.
     $$
     Integrating both sides:
     $$
     \int 2y \, dy = \int dx \quad \Rightarrow \quad y^2 = x + C.
     $$
     Taking the square root:
     $$
     y(x) = \pm \sqrt{x + C}.
     $$

## 2. Solve the first-order ordinary differential equations using the method of separation of variables for $x$ and $y=y(x)$:

   - $\frac{dy}{dx} = \frac{x}{y}$

     We can separate the variables:
     $$
     y \, dy = x \, dx.
     $$
     Integrating both sides:
     $$
     \int y \, dy = \int x \, dx \quad \Rightarrow \quad \frac{y^2}{2} = \frac{x^2}{2} + C.
     $$
     Simplifying:
     $$
     y^2 = x^2 + C.
     $$

   - $\frac{dy}{dx} = \frac{y}{x}$

     We can separate the variables:
     $$
     \frac{dy}{y} = \frac{dx}{x}.
     $$
     Integrating both sides:
     $$
     \int \frac{1}{y} \, dy = \int \frac{1}{x} \, dx \quad \Rightarrow \quad \ln|y| = \ln|x| + C.
     $$
     Exponentiating both sides:
     $$
     y(x) = Cx.
     $$

   - $\frac{dy}{dx} = xy$

     We can separate the variables:
     $$
     \frac{dy}{y} = x \, dx.
     $$
     Integrating both sides:
     $$
     \int \frac{1}{y} \, dy = \int x \, dx \quad \Rightarrow \quad \ln|y| = \frac{x^2}{2} + C.
     $$
     Exponentiating both sides:
     $$
     y(x) = Ce^{\frac{x^2}{2}}.
     $$

## 3. Solve the second-order ordinary differential equations:

   * $y''(x) + y'(x) = 0$, with boundary conditions $y(0) = 2$ and $y'(0) = -1$

     This is a linear first-order equation for \( y'(x) \). We can solve it as:
     $$
     y'(x) = Ce^{-x}.
     $$
     Integrating:
     $$
     y(x) = -Ce^{-x} + D.
     $$
     Using the initial conditions \( y(0) = 2 \) and \( y'(0) = -1 \):
     $$
     y(0) = -C + D = 2 \quad \Rightarrow \quad D = C + 2,
     $$
     and
     $$
     y'(0) = -C = -1 \quad \Rightarrow \quad C = 1.
     $$
     Therefore, the solution is:
     $$
     y(x) = -e^{-x} + 3.
     $$

   * $y''(x) - y(x)= 0$, with boundary conditions $y(0) = 2$ and $y'(0) = 0$

     This is a homogeneous second-order differential equation. The characteristic equation is:
     $$
     r^2 - 1 = 0 \quad \Rightarrow \quad r = \pm 1.
     $$
     Therefore, the general solution is:
     $$
     y(x) = A e^x + B e^{-x}.
     $$
     Using the initial conditions:
     $$
     y(0) = A + B = 2 \quad \text{and} \quad y'(0) = A - B = 0,
     $$
     we find \( A = B = 1 \). Thus, the solution is:
     $$
     y(x) = e^x + e^{-x} = 2 \cosh(x).
     $$

   * $\frac{d^2 y(x)}{dx^2} = -\omega^2 y(x)$.

     This is a standard equation for simple harmonic motion. The characteristic equation is:
     $$
     r^2 + \omega^2 = 0 \quad \Rightarrow \quad r = \pm i \omega.
     $$
     Therefore, the general solution is:
     $$
     y(x) = A \cos(\omega x) + B \sin(\omega x).
     $$

## 4. Check if the function $\psi(t, x) = A \cos(\omega t + kx)$ is a solution of the second-order partial differential equation (the so-called "wave equation"), where $v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda}$:

   The wave equation is:
   $$
   \frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0,
   $$
   where \( v = \frac{\omega}{k} \).

   First, calculate the second partial derivatives:

   $$
   \frac{\partial^2 \psi(t, x)}{\partial t^2} = -A \omega^2 \cos(\omega t + kx),
   $$
   and
   $$
   \frac{\partial^2 \psi(t, x)}{\partial x^2} = -A k^2 \cos(\omega t + kx).
   $$

   Substituting these into the wave equation:
   $$
   -A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx)) = 0.
   $$
   Simplifying:
   $$
   -A \omega^2 \cos(\omega t + kx) + A v^2 k^2 \cos(\omega t + kx) = 0.
   $$
   Since \( v = \frac{\omega}{k} \), we have:
   $$
   v^2 k^2 = \omega^2,
   $$
   so the equation simplifies to:
   $$
   0 = 0.
   $$
   Therefore, \( \psi(t, x) = A \cos(\omega t + kx) \) is indeed a solution to the wave equation.
