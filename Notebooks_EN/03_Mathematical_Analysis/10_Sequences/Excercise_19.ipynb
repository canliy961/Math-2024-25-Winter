### 1. Calculate:

#### (a) $\displaystyle \lim_{n \to \infty} \frac{n^2 + 3n}{2 n^2 - 2n}$

Factor out $n^2$ from numerator and denominator:

$$ \frac{n^2 + 3n}{2n^2 - 2n} = \frac{n^2(1 + 3/n)}{n^2(2 - 2/n)} = \frac{1 + 3/n}{2 - 2/n}. $$

As $n \to \infty$, the terms $3/n$ and $2/n$ approach 0, so:

$$ \lim_{n \to \infty} \frac{1 + 3/n}{2 - 2/n} = \frac{1 + 0}{2 - 0} = \frac{1}{2}. $$

#### (b) $\displaystyle \lim_{n \to \infty} \frac{(2n+3)^3}{n^3 - 1}$

Expand (or factor out $n^3$ directly):

$$(2n+3)^3 = 8n^3 + 36n^2 + 54n + 27. $$

Then:

$$ \frac{(2n+3)^3}{n^3 - 1} = \frac{8n^3 + 36n^2 + 54n + 27}{n^3 - 1}. $$

Divide numerator and denominator by $n^3$:

$$ \frac{8 + 36/n + 54/n^2 + 27/n^3}{1 - 1/n^3}. $$

As $n \to \infty$, the terms $36/n, 54/n^2, 27/n^3,$ and $1/n^3$ all go to 0, leaving:

$$ 8. $$

### 2. Prove using the squeeze theorem:

#### $\displaystyle \lim_{n \to \infty} \frac{\sin(n)}{n}$

For all integers $n$, we know:

$$ -1 \leq \sin(n) \leq 1. $$

Divide each part by $n > 0$ (for large $n$):

$$ \frac{-1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}. $$

As $n \to \infty$, $-1/n \to 0$ and $1/n \to 0$. By the squeeze theorem:

$$ \lim_{n \to \infty} \frac{\sin(n)}{n} = 0. $$

It is well-known that:

$$ \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e. $$


### 2. Prove using the squeeze theorem:
$$
\lim_{n \to \infty} \frac{\sin(n)}{n}
$$

---

### Solution:

We want to prove the following limit using the **Squeeze Theorem**:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n}
$$

#### Step 1: Apply the known inequality for \( \sin(n) \)

We know that for all real numbers \( n \), the sine function satisfies the inequality:
$$
-1 \leq \sin(n) \leq 1
$$

#### Step 2: Divide through by \( n \)

For \( n > 0 \), we divide the inequality by \( n \) to obtain:
$$
\frac{-1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}
$$

#### Step 3: Take the limit as \( n \to \infty \)

Now, take the limit of each part of the inequality as \( n \to \infty \):

- The limit of \( \frac{-1}{n} \) as \( n \to \infty \) is 0:
  $$
  \lim_{n \to \infty} \frac{-1}{n} = 0
  $$

- The limit of \( \frac{1}{n} \) as \( n \to \infty \) is also 0:
  $$
  \lim_{n \to \infty} \frac{1}{n} = 0
  $$

Thus, we have:
$$
0 \leq \lim_{n \to \infty} \frac{\sin(n)}{n} \leq 0
$$

#### Step 4: Apply the squeeze theorem

Since \( \frac{\sin(n)}{n} \) is squeezed between two expressions that both approach 0, by the **Squeeze Theorem**, we conclude that:
$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0
$$

Thus, the limit is:
$$
\boxed{0}
$$


### 3. Find the limit of the sequence:
$$
a_n = \left( 1 + \frac{1}{n} \right)^n
$$

This is a well-known limit, and it approaches the mathematical constant \( e \). Let's prove this explicitly.

**Solution:**

To find the limit of \( a_n \), we can use logarithms. First, take the natural logarithm of both sides:
$$
\ln(a_n) = \ln \left( \left( 1 + \frac{1}{n} \right)^n \right)
$$
By the logarithmic property \( \ln(x^y) = y \ln(x) \), we can simplify the right-hand side:
$$
\ln(a_n) = n \ln \left( 1 + \frac{1}{n} \right)
$$

Now, to find the limit as \( n \to \infty \), we need to examine the term \( \ln \left( 1 + \frac{1}{n} \right) \) for large values of \( n \).

**Step 1: Approximate \( \ln \left( 1 + \frac{1}{n} \right) \) for large \( n \)**

For large \( n \), we can use the first-order Taylor expansion of \( \ln(1 + x) \) around \( x = 0 \):
$$
\ln(1 + x) \approx x \quad \text{when} \quad x \text{ is small}
$$
In our case, \( x = \frac{1}{n} \), so for large \( n \), we have:
$$
\ln \left( 1 + \frac{1}{n} \right) \approx \frac{1}{n}
$$

**Step 2: Substitute the approximation into the expression for \( \ln(a_n) \)**

Now substitute this approximation into the expression for \( \ln(a_n) \):
$$
\ln(a_n) = n \ln \left( 1 + \frac{1}{n} \right) \approx n \cdot \frac{1}{n} = 1
$$
Thus, as \( n \to \infty \), \( \ln(a_n) \to 1 \).

**Step 3: Exponentiate both sides**

Since \( \ln(a_n) \to 1 \) as \( n \to \infty \), we exponentiate both sides to find the limit of \( a_n \):
$$
a_n \to e^1 = e
$$

**Conclusion:**

Thus, the limit of the sequence is:
$$
\lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n = e
$$
Therefore, the limit is:
$$
\boxed{e}
$$
