# 1. Find the Taylor series and visualize obtained functions in Geogebra:
   - $f(x) = \cos(x)$ around $x = 0$ up to the 4th degree.
   - $h(x) = \frac{1}{1-x}$ around $x = 0$ up to the 4th degree.
   - $g(x) = \sin(x)$ around $x = \pi$ up to the 4th degree.

### For $f(x) = \cos(x)$ around $x = 0$ up to the 4th degree:

The Taylor series for $\cos(x)$ around $x = 0$ is given by:

$$
\cos(x) = \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n)!} x^{2n}
$$

Up to the 4th degree, the series becomes:

$$
\cos(x) \approx 1 - \frac{x^2}{2!} + \frac{x^4}{4!} = 1 - \frac{x^2}{2} + \frac{x^4}{24}
$$

Thus, the Taylor series approximation of $\cos(x)$ around $x = 0$ up to the 4th degree is:

$$
\cos(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24}
$$

### For $h(x) = \frac{1}{1-x}$ around $x = 0$ up to the 4th degree:

The Taylor series for $\frac{1}{1-x}$ around $x = 0$ is:

$$
\frac{1}{1-x} = \sum_{n=0}^{\infty} x^n
$$

Up to the 4th degree, the series becomes:

$$
\frac{1}{1-x} \approx 1 + x + x^2 + x^3 + x^4
$$

Thus, the Taylor series approximation of $\frac{1}{1-x}$ around $x = 0$ up to the 4th degree is:

$$
\frac{1}{1-x} \approx 1 + x + x^2 + x^3 + x^4
$$

### For $g(x) = \sin(x)$ around $x = \pi$ up to the 4th degree:

The Taylor series for $\sin(x)$ around $x = \pi$ is:

$$
\sin(x) = \sin(\pi) + \cos(\pi)(x - \pi) - \frac{\sin(\pi)}{2!}(x - \pi)^2 - \frac{\cos(\pi)}{3!}(x - \pi)^3 + \frac{\sin(\pi)}{4!}(x - \pi)^4 + \dots
$$

Since $\sin(\pi) = 0$ and $\cos(\pi) = -1$, this simplifies to:

$$
\sin(x) \approx - (x - \pi) + \frac{(x - \pi)^3}{6}
$$

Thus, the Taylor series approximation of $\sin(x)$ around $x = \pi$ up to the 4th degree is:

$$
\sin(x) \approx - (x - \pi) + \frac{(x - \pi)^3}{6}
$$

# 2. Find a tangent line $y = f'(x_0) (x-x_0) + f(x_0)$ to the function $f(x) = e^{\sin(x)}$ at $x_0 = \pi$.

### First, compute $f'(x)$ for $f(x) = e^{\sin(x)}$:

Using the chain rule:

$$
f'(x) = e^{\sin(x)} \cos(x)
$$

### Now, evaluate $f(x)$ and $f'(x)$ at $x_0 = \pi$:

Since $\sin(\pi) = 0$ and $\cos(\pi) = -1$, we have:

$$
f(\pi) = e^{\sin(\pi)} = e^0 = 1
$$

$$
f'(\pi) = e^{\sin(\pi)} \cos(\pi) = 1 \times (-1) = -1
$$

Thus, the equation of the tangent line at $x_0 = \pi$ is:

$$
y = f'(\pi) (x - \pi) + f(\pi) = -1(x - \pi) + 1
$$

Simplifying the expression:

$$
y = -x + \pi + 1
$$

Thus, the equation of the tangent line to the function $f(x) = e^{\sin(x)}$ at $x_0 = \pi$ is:

$$
y = -x + \pi + 1
$$
