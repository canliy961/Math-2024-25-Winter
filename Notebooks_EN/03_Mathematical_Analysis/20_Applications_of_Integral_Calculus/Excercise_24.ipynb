## 1. Compute the following integrals:

$$
\int (x^2 + 2) \, dx = \frac{x^3}{3} + 2x + C
$$
$$
\int 2 \sin(x) \, dx = -2 \cos(x) + C

$$
\int 3x \, dx = 3 \ln |x| + C

$$
\int \frac{1}{x^2} \, dx = -\frac{1}{x} + C
$$
$$
\int \left( \frac{1}{3} x^4 - 5 \right) \, dx = \frac{x^5}{15} - 5x + C

$$
\int (\sin^2 x + \cos^2 x) \, dx = \int 1 \, dx = x + C \quad (\text{since} \ \sin^2(x) + \cos^2(x) = 1)

$$
\int (5 \sin(x) + 3 e^x) \, dx = -5 \cos(x) + 3 e^x + C

## 2. Compute the integrals over the interval $[0, \pi]$:
- For $f(x) = 2x + 1$:
$$
\int_0^\pi (2x + 1) \, dx = \left[ x^2 + x \right]_0^\pi = \left( \pi^2 + \pi \right) - 0 = \pi^2 + \pi
$$
- For $g(x) = x^2$:
$$
\int_0^\pi x^2 \, dx = \left[ \frac{x^3}{3} \right]_0^\pi = \frac{\pi^3}{3} - 0 = \frac{\pi^3}{3}
$$

## 3. Compute the area of the region bounded by the curves:
Given the region:
$$
x = 1, \quad x = 2, \quad y = 0, \quad y = x^2 + 1
$$
The area is calculated as:
$$
A = \int_1^2 (x^2 + 1) \, dx = \left[ \frac{x^3}{3} + x \right]_1^2
$$
$$
A = \left( \frac{8}{3} + 2 \right) - \left( \frac{1}{3} + 1 \right) = \frac{8}{3} + 2 - \frac{1}{3} - 1 = \frac{7}{3} + 1 = \frac{10}{3}
$$

## 4. Calculate the area under the sine curve over the interval $[0, \pi]$:
The given integral is:
$$
P = \int_0^\pi \sin(x) \, dx = -\cos(x) \bigg|_0^\pi = -\cos(\pi) + \cos(0) = 1 + 1 = 2
$$

## 6. Find the distance traveled by the moving particle between $t = 0$ and $t = 2$ for the position function $x(t) = 3t^2 - 6t + 1$:
First, compute the velocity function:
$$
v(t) = \frac{dx}{dt} = 6t - 6
$$
The total distance is the integral of the absolute value of the velocity:
$$
\text{Distance} = \int_0^2 |6t - 6| \, dt
$$
The velocity function changes sign at $t = 1$, so we split the integral into two parts:
$$
\int_0^2 |6t - 6| \, dt = \int_0^1 -(6t - 6) \, dt + \int_1^2 (6t - 6) \, dt
$$
Solving the first part:
$$
\int_0^1 -(6t - 6) \, dt = \int_0^1 (-6t + 6) \, dt = \left[ -3t^2 + 6t \right]_0^1 = (-3 + 6) - 0 = 3
$$
Solving the second part:
$$
\int_1^2 (6t - 6) \, dt = \int_1^2 (6t - 6) \, dt = \left[ 3t^2 - 6t \right]_1^2 = (12 - 12) - (3 - 6) = 3
$$
Total distance:
$$
\text{Distance} = 3 + 3 = 6
$$
