### 1. Compute:

1. $\displaystyle \int 1 \, dx$

$$\int 1 \, dx = x + C.$$

2. $\displaystyle \int (x^2 + 2) \, dx$

$$\int (x^2 + 2) \, dx = \frac{x^3}{3} + 2x + C.$$

3. $\displaystyle \int 2 \sin(x) \, dx$

$$\int 2 \sin(x) \, dx = -2 \cos(x) + C.$$

4. $\displaystyle \int \frac{3}{x} \, dx$

$$\int \frac{3}{x} \, dx = 3 \ln|x| + C.$$

5. $\displaystyle \int \frac{1}{x^2} \, dx$

Rewrite $\frac{1}{x^2}$ as $x^{-2}$:

$$\int x^{-2} \, dx = \int x^{-2} \, dx = -x^{-1} + C = -\frac{1}{x} + C.$$

6. $\displaystyle \int \left( \frac{1}{3} x^4 - 5 \right) \, dx$

$$\int \left( \frac{1}{3} x^4 - 5 \right) \, dx = \frac{1}{3} \cdot \frac{x^5}{5} - 5x + C = \frac{x^5}{15} - 5x + C.$$

7. $\displaystyle \int (\sin^2 x + \cos^2 x) \, dx$

Since $\sin^2 x + \cos^2 x = 1$:

$$\int (\sin^2 x + \cos^2 x) \ dx = \int 1 \ dx = x + C.$$

8. $\displaystyle \int (5 \sin x + 3 e^x) \, dx$

$$\int 5 \sin x \ dx + \int 3 e^x \ dx = -5 \cos x + 3 e^x + C.$$

9. $\displaystyle \int \sqrt[3]{x} \, dx = \int x^{1/3} \, dx$

$$\int x^{1/3} \, dx = \frac{x^{4/3}}{\frac{4}{3}} + C = \frac{3}{4} x^{4/3} + C.$$

10. $\displaystyle \int \sqrt{10x} \ dx = \int (10x)^{1/2} \ dx = \int 10^{1/2} x^{1/2} \ dx$

Factor out $\sqrt{10}$:

$$\sqrt{10} \int x^{1/2} \ dx = \sqrt{10} \cdot \frac{2}{3} x^{3/2} + C = \frac{2 \sqrt{10}}{3} x^{3/2} + C.$$

11. $\displaystyle \int \cos\Bigl(\frac{5}{2}x + 3\Bigr) \ dx$

Let $u = \frac{5}{2}x + 3$, then $du = \frac{5}{2} dx$, or $dx = \frac{2}{5} du$.

$$\int \cos(u) \cdot \frac{2}{5} \, du = \frac{2}{5} \int \cos(u) \, du = \frac{2}{5} \sin(u) + C = \frac{2}{5} \sin\Bigl(\frac{5}{2}x + 3\Bigr) + C.$$

12. $\displaystyle \int \frac{\cos(\ln(x))}{x} \ dx$

Let $u = \ln(x)$, then $du = \frac{1}{x} dx$ so $dx = x\,du$, but also the factor $\frac{1}{x} dx$ is just $du$.

Hence:

$$\int \frac{\cos(\ln(x))}{x} \ dx = \int \cos(u) \, du = \sin(u) + C = \sin(\ln(x)) + C.$$

13. $\displaystyle \int x \ln(x) \, dx$

Use integration by parts:

Let $u = \ln(x)$, $dv = x \, dx$. Then $du = \frac{1}{x} \, dx$ and $v = \frac{x^2}{2}$.

$$\int x \ln(x) \, dx = \left(\ln(x)\right)\left(\frac{x^2}{2}\right) - \int \frac{x^2}{2} \cdot \frac{1}{x} \ dx = \frac{x^2}{2}\ln(x) - \int \frac{x}{2} \ dx.$$

$$= \frac{x^2}{2} \ln(x) - \frac{1}{2} \int x \ dx = \frac{x^2}{2} \ln(x) - \frac{1}{2} \cdot \frac{x^2}{2} + C.$$

$$= \frac{x^2}{2} \ln(x) - \frac{x^2}{4} + C.$$

14. $\displaystyle \int x e^x \, dx$

Use integration by parts again:

Let $u = x$, $dv = e^x dx$. Then $du = dx$, $v = e^x$.

$$\int x e^x \ dx = x e^x - \int e^x \ dx = x e^x - e^x + C = e^x (x - 1) + C.$$

### 2. Calculate integrals over the interval $[0, \pi]$ and visualize them in Geogebra:

1. $f(x) = 2x + 1$

$$\int_0^{\pi} (2x + 1) \ dx = \left[ x^2 + x \right]_0^{\pi} = (\pi^2 + \pi) - (0 + 0) = \pi^2 + \pi.$$

2. $g(x) = x^2$

$$\int_0^{\pi} x^2 \ dx = \left[ \frac{x^3}{3} \right]_0^{\pi} = \frac{\pi^3}{3} - 0 = \frac{\pi^3}{3}.$$

### 3. Calculate the area of the region bounded by the lines:

- $x = 1$, $x = 2$, $y = 0$, and $y = x^2 + 1$.

The area $A$ is given by the definite integral of the top curve $y = x^2 + 1$ from $x=1$ to $x=2$, minus the line $y=0$ (which is just 0):

$$A = \int_{1}^{2} (x^2 + 1) \, dx = \left[ \frac{x^3}{3} + x \right]_1^2.$$

Evaluate:

- At $x=2$: $\frac{(2)^3}{3} + 2 = \frac{8}{3} + 2 = \frac{8}{3} + \frac{6}{3} = \frac{14}{3}$.
- At $x=1$: $\frac{1}{3} + 1 = \frac{4}{3}$.

So:

$$A = \frac{14}{3} - \frac{4}{3} = \frac{10}{3}.$$

### 4. Calculate the area under the sine curve over the interval $[0, \pi]$:

$$P = \int_0^{\pi} \sin(x) \, dx.$$

Evaluate:

$$\int_0^{\pi} \sin(x) \ dx = \left[ -\cos(x) \right]_0^{\pi} = [-\cos(\pi)] - [-\cos(0)] = -(-1) - [-1] = 1 + 1 = 2.$$

### 5. Calculate the length of the sine curve over the interval $[0, \pi]$:

$$L = \int_0^{\pi} \sqrt{1 + (f'(x))^2} \ dx = \int_0^{\pi} \sqrt{1 + \cos^2(x)} \ dx.$$

This integral does not have a simple closed-form expression in elementary functions. It can be expressed in terms of elliptic integrals. Numerically, one can approximate it or use special functions.

Hence:

$$L = \int_0^{\pi} \sqrt{1 + \cos^2(x)} \ dx \approx 3.8201977891\ldots\,.$$

### 6. Find the distance of the moving particle between time $t=0$ and $t=2$ for the position function $x(t) = 3t^2 - 6t + 1$.

First, find the velocity:

$$v(t) = x'(t) = 6t - 6.$$

The distance traveled (assuming velocity could change sign) is:

$$\text{Distance} = \int_{0}^{2} |v(t)| \ dt = \int_{0}^{2} |6t - 6| \ dt.$$

We check the sign of $6t-6$ on $[0,2]$:

- $6t-6 = 6(t-1)$. This is 0 at $t=1$.
- For $0 \le t < 1$, $6t - 6 < 0$.
- For $1 < t \le 2$, $6t - 6 > 0$.

So we split the integral:

$$\text{Distance} = \int_{0}^{1} [-(6t-6)] \ dt + \int_{1}^{2} [(6t-6)] \ dt,$$

because for $t<1$, $6t-6$ is negative, so its absolute value is $-(6t-6)$, and for $t>1$, $6t-6$ is positive.

**First part:**

$$\int_{0}^{1} [6 - 6t] \ dt = \left[ 6t - 3t^2 \right]_0^1 = (6 - 3) - (0) = 3.$$

**Second part:**

$$\int_{1}^{2} [6t - 6] \ dt = \left[ 3t^2 - 6t \right]_1^2.$$

- At $t=2$: $3(4) - 6(2) = 12 - 12 = 0$.
- At $t=1$: $3(1) - 6(1) = 3 - 6 = -3$.

So this integral is $0 - (-3) = 3.$

Hence the total distance is:

$$3 + 3 = 6.$$
