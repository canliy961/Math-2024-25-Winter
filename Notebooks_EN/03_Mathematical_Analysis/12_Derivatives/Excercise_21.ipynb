### 1. Compute derivatives of the following functions:

1. $y(x) = -3x + 3$

$$\frac{d}{dx} [-3x + 3] = -3.$$

2. $y(x) = \pi x + \sin(1)$

$$\frac{d}{dx}[\pi x + \sin(1)] = \pi,$$

since $\sin(1)$ is a constant with respect to $x$.

3. $y(x) = 4 + \sin(2)$

$$\frac{d}{dx}[4 + \sin(2)] = 0,$$

since it is a constant.

4. $y(x) = 2x^3 - 3x^2 + 8x - 9$

$$\frac{d}{dx}[2x^3 - 3x^2 + 8x - 9] = 6x^2 - 6x + 8.$$

5. $y(x) = 6 x^{1/3}$

$$\frac{d}{dx}[6 x^{1/3}] = 6 \cdot \frac{1}{3} x^{-2/3} = 2 x^{-2/3} = \frac{2}{x^{2/3}}.$$

6. $y(x) = \sqrt{x} = x^{1/2}$

$$\frac{d}{dx}[x^{1/2}] = \frac{1}{2} x^{-1/2} = \frac{1}{2\sqrt{x}}.$$

7. $y(x) = \cos(x) + \sin(x)$

$$\frac{d}{dx}[\cos(x) + \sin(x)] = -\sin(x) + \cos(x).$$

8. $y(x) = 2\sin(x) \cos(x)$

Using product rule or the identity $\sin(2x) = 2\sin(x)\cos(x)$:

- **Method A (product rule):**

  $$\frac{d}{dx}[2\sin(x)\cos(x)] = 2[\cos(x)\cos(x) - \sin(x)\sin(x)] = 2[\cos^2(x) - \sin^2(x)] = 2\cos(2x).$$

- **Method B (double-angle identity):**

  $$2\sin(x)\cos(x) = \sin(2x),$$

  $$\frac{d}{dx}[\sin(2x)] = 2\cos(2x).$$

Either way, the derivative is $2\cos(2x)$.

9. $y(x) = x\sin(x)$

By the product rule:

$$\frac{d}{dx}[x\sin(x)] = x\cos(x) + \sin(x).$$

10. $y(x) = (x+1)(x+1) = (x+1)^2$

$$\frac{d}{dx}[(x+1)^2] = 2(x+1).$$

11. $y(x) = \frac{x}{x+1}$

Use the quotient rule:

$$\frac{d}{dx}\left[\frac{x}{x+1}\right] = \frac{(x+1)\cdot 1 - x\cdot 1}{(x+1)^2} = \frac{x+1 - x}{(x+1)^2} = \frac{1}{(x+1)^2}.$$

12. $y(x) = (x+1)\exp(x)$

By the product rule:

$$\frac{d}{dx}[(x+1) e^x] = (x+1) e^x + 1 \cdot e^x = (x+2)e^x.$$

13. $y(x) = \sin(x^2)$

By the chain rule:

$$\frac{d}{dx}[\sin(x^2)] = \cos(x^2) \cdot 2x = 2x\cos(x^2).$$

14. $y(x) = \exp(-2x)$

By the chain rule:

$$\frac{d}{dx}[e^{-2x}] = -2 e^{-2x}.$$

15. $y(x) = \frac{1}{\sin(x+1)} = \sin(x+1)^{-1}$

By the chain rule:

$$\frac{d}{dx}[\sin(x+1)^{-1}] = -1 \cdot \sin(x+1)^{-2} \cdot \cos(x+1) = -\frac{\cos(x+1)}{\sin^2(x+1)}.$$

16. $y(x) = \sqrt{2x+1} = (2x+1)^{1/2}$

By the chain rule:

$$\frac{d}{dx}[(2x+1)^{1/2}] = \frac{1}{2}(2x+1)^{-1/2} \cdot 2 = \frac{1}{\sqrt{2x+1}}.$$

### 2. Prove:
$$\frac{d}{dx} [\ln(\sin(x))] = \cot(x).$$

We use the chain rule:

$$\frac{d}{dx}[\ln(\sin(x))] = \frac{1}{\sin(x)} \cdot \cos(x) = \frac{\cos(x)}{\sin(x)} = \cot(x).$$

$$f''(x) = \frac{d}{dx}[-\sin(x)] = -\cos(x).$$

### 4. Using de l'Hospital's Rule, find the improper limits:

1. $\displaystyle \lim_{x\to 0} \frac{\sin{x}}{x}$

Directly, as $x\to 0$, $\sin(x)\sim x$. One can also use de l'Hospital's rule:

- Numerator $\sin(x)$ and denominator $x$ both go to 0.
- Differentiate top and bottom:

  $$\lim_{x\to 0} \frac{\sin{x}}{x} = \lim_{x\to 0} \frac{\cos(x)}{1} = 1.$$

2. $\displaystyle \lim_{x\to \infty} \frac{\ln x}{x}$

As $x\to \infty$, $\ln x\to \infty$ and $x\to \infty$. Apply de l'Hospital's rule:

- Differentiate top and bottom:

  $$\lim_{x\to \infty} \frac{\ln x}{x} = \lim_{x\to \infty} \frac{\frac{1}{x}}{1} = 0.$$

3. $\displaystyle \lim_{x\to \infty} \frac{\exp(x)}{x}$

As $x\to \infty$, $e^x\to \infty$ and $x\to \infty$. Again, apply de l'Hospital's rule:

- Differentiate top and bottom:

  $$\lim_{x\to \infty} \frac{e^x}{x} = \lim_{x\to \infty} \frac{e^x}{1} = \infty.$$

### 2. Prove

$$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \cot(x)
$$

### Proof:

We will differentiate \( \ln(\sin(x)) \) using the chain rule.

Let \( y = \ln(\sin(x)) \). To differentiate, we apply the chain rule, which states:

$$
\frac{d}{dx} \ln(u) = \frac{1}{u} \cdot \frac{du}{dx}
$$

In this case, \( u = \sin(x) \). Therefore, we have:

$$
\frac{d}{dx} \ln(\sin(x)) = \frac{1}{\sin(x)} \cdot \frac{d}{dx} \sin(x)
$$

Now, we differentiate \( \sin(x) \):

$$
\frac{d}{dx} \sin(x) = \cos(x)
$$

Thus, the derivative becomes:

$$
\frac{d}{dx} \ln(\sin(x)) = \frac{1}{\sin(x)} \cdot \cos(x)
$$

Simplifying:

$$
\frac{\cos(x)}{\sin(x)} = \cot(x)
$$

Therefore, we have:

$$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \cot(x)
$$

### 4. Using de L'Hopital's Rule, find the following improper limits:

(a)

$$ \lim_{x \to 0} \frac{\sin(x)}{x} $$

(b)

$$ \lim_{x \to \infty} \frac{\ln(x)}{x} $$

(c)

$$ \lim_{x \to \infty} \frac{\exp(x)}{x} $$

---

### Solution:

#### (a) Limit of \( \frac{\sin(x)}{x} \) as \( x \to 0 \):

We are given the limit:

$$ \lim_{x \to 0} \frac{\sin(x)}{x} $$

This is an indeterminate form \( \frac{0}{0} \), so we apply L'Hopital's Rule. Differentiate the numerator and the denominator:

- The derivative of \( \sin(x) \) is \( \cos(x) \),
- The derivative of \( x \) is \( 1 \).

Thus, applying L'Hopital's Rule:

$$ \lim_{x \to 0} \frac{\sin(x)}{x} = \lim_{x \to 0} \frac{\cos(x)}{1} $$

Since \( \cos(0) = 1 \), we have:

$$ \lim_{x \to 0} \frac{\sin(x)}{x} = 1 $$

#### (b) Limit of \( \frac{\ln(x)}{x} \) as \( x \to \infty \):

Now consider the limit:

$$ \lim_{x \to \infty} \frac{\ln(x)}{x} $$

This is also an indeterminate form \( \frac{\infty}{\infty} \), so we apply L'Hopital's Rule. Differentiate the numerator and the denominator:

- The derivative of \( \ln(x) \) is \( \frac{1}{x} \),
- The derivative of \( x \) is \( 1 \).

Thus, applying L'Hopital's Rule:

$$ \lim_{x \to \infty} \frac{\ln(x)}{x} = \lim_{x \to \infty} \frac{\frac{1}{x}}{1} = \lim_{x \to \infty} \frac{1}{x} $$

As \( x \to \infty \), \( \frac{1}{x} \to 0 \), so we have:

$$ \lim_{x \to \infty} \frac{\ln(x)}{x} = 0 $$

#### (c) Limit of \( \frac{\exp(x)}{x} \) as \( x \to \infty \):

Finally, consider the limit:

$$ \lim_{x \to \infty} \frac{\exp(x)}{x} $$

This is an indeterminate form \( \frac{\infty}{\infty} \), so we apply L'Hopital's Rule. Differentiate the numerator and the denominator:

- The derivative of \( \exp(x) \) is \( \exp(x) \),
- The derivative of \( x \) is \( 1 \).

Thus, applying L'Hopital's Rule:

$$ \lim_{x \to \infty} \frac{\exp(x)}{x} = \lim_{x \to \infty} \frac{\exp(x)}{1} = \lim_{x \to \infty} \exp(x) $$

Since \( \exp(x) \to \infty \) as \( x \to \infty \), we have:

$$ \lim_{x \to \infty} \frac{\exp(x)}{x} = \infty $$

---

### Final Answers:

(a)

$$ \lim_{x \to 0} \frac{\sin(x)}{x} = 1 $$

(b)

$$ \lim_{x \to \infty} \frac{\ln(x)}{x} = 0 $$

(c)

$$ \lim_{x \to \infty} \frac{\exp(x)}{x} = \infty $$



### 5. In physics, the position of a particle is given by:

$$ x(t) = 3t^2 - 6t + 1 $$

Find the velocity \( V(t) = x'(t) \) and acceleration \( a(t) = V'(t) = x''(t) \) of the particle at time \( t = 2 \).

---

### Solution:

#### 1. Velocity:

The velocity \( V(t) \) is the first derivative of the position function \( x(t) \) with respect to \( t \).

Differentiate \( x(t) = 3t^2 - 6t + 1 \) with respect to \( t \):

$$ V(t) = x'(t) = \frac{d}{dt}(3t^2 - 6t + 1) $$

Using the power rule:

$$ V(t) = 6t - 6 $$

Now, evaluate the velocity at \( t = 2 \):

$$ V(2) = 6(2) - 6 = 12 - 6 = 6 $$

Thus, the velocity at \( t = 2 \) is:

$$ V(2) = 6 $$

#### 2. Acceleration:

The acceleration \( a(t) \) is the derivative of the velocity function \( V(t) \), or equivalently, the second derivative of the position function \( x(t) \).

Now, differentiate \( V(t) = 6t - 6 \) with respect to \( t \):

$$ a(t) = V'(t) = \frac{d}{dt}(6t - 6) = 6 $$

So, the acceleration is constant:

$$ a(t) = 6 $$

Thus, the acceleration at \( t = 2 \) is:

$$ a(2) = 6 $$

---

### Final Answer:

At \( t = 2 \), the velocity and acceleration are:

$$ V(2) = 6, \quad a(2) = 6 $$

Therefore, the velocity and acceleration of the particle at \( t = 2 \) are both:

$$ \boxed{6} $$
