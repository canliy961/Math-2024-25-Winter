### 6. The profit function is $P(u) = -2u^2 + 50u - 300$, where $u$ is the number of units sold. Find the number of units that maximize profit.

To find the extremum, we first find the derivative of $P(u)$:

$$
P'(u) = \frac{d}{du}(-2u^2 + 50u - 300) = -4u + 50
$$

Set the derivative equal to zero to find the critical points:

$$
-4u + 50 = 0
$$

Solving for $u$:

$$
4u = 50 \quad \Rightarrow \quad u = \frac{50}{4} = 12.5
$$

To confirm that this critical point corresponds to a maximum, we check the second derivative:

$$
P''(u) = \frac{d}{du}(-4u + 50) = -4
$$

Since $P''(u) = -4$ is negative, the function has a maximum at $u = 12.5$.

Thus, the number of units that maximize profit is $12.5$.

### 7. You have 10 meters of string, and you need to use it to enclose the largest possible rectangular. Find the dimensions of the rectangle.

Let the length of the rectangle be $L$ meters and the width be $W$ meters. The perimeter constraint is:

$$
2L + 2W = 10
$$

Simplifying the equation:

$$
L + W = 5 \quad \Rightarrow \quad W = 5 - L
$$

The area of the rectangle is:

$$
A(L) = L \cdot W = L(5 - L) = 5L - L^2
$$

To maximize the area, we take the derivative of $A(L)$:

$$
A'(L) = \frac{d}{dL}(5L - L^2) = 5 - 2L
$$

Set the derivative equal to zero:

$$
5 - 2L = 0 \quad \Rightarrow \quad L = \frac{5}{2} = 2.5
$$

Substitute $L = 2.5$ into the perimeter equation:

$$
W = 5 - L = 5 - 2.5 = 2.5
$$

Thus, the dimensions of the rectangle that maximize the area are $2.5$ meters by $2.5$ meters.

### 8. Find extremum of $f(x) = x^2 + 3x - 5$.

To find the extremum, we first find the derivative of $f(x)$:

$$
f'(x) = \frac{d}{dx}(x^2 + 3x - 5) = 2x + 3
$$

Set the derivative equal to zero to find the critical points:

$$
2x + 3 = 0 \quad \Rightarrow \quad x = -\frac{3}{2}
$$

To confirm whether this is a minimum or maximum, we check the second derivative:

$$
f''(x) = \frac{d}{dx}(2x + 3) = 2
$$

Since $f''(x) = 2 > 0$, the function has a minimum at $x = -\frac{3}{2}$.

Thus, the function has a minimum at $x = -\frac{3}{2}$.

### 9. Find extremum of $f(x) = \frac{x^2 + 2x + 1}{x - 1}$.

$$
f'(x) = \frac{h(x)g'(x) - g(x)h'(x)}{(h(x))^2}
$$

$$
g'(x) = 2x + 2, \quad h'(x) = 1
$$

Now, apply the quotient rule:

$$

= \frac{(2x^2 + 2x - 2x - 2) - (x^2 + 2x + 1)}{(x - 1)^2}

$$
= \frac{2x^2 - 2 - x^2 - 2x - 1}{(x - 1)^2}

$$
= \frac{x^2 - 2x - 3}{(x - 1)^2}

Set the derivative equal to zero to find the critical points:

$$
x^2 - 2x - 3 = 0

$$

Thus, $x = 3$ or $x = -1$. We need to check whether these are local minima or maxima. Since $f(x)$ is undefined at $x = 1$, we exclude this point.

Thus, the critical points are $x = 3$ and $x = -1$.
