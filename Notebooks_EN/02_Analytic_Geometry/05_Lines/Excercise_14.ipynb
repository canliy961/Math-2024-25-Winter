### 1. The line passes through points \(A(1, 2)\) and \(B(3, 4)\). Find the equation of the line.

To find the equation of the line passing through points \(A(1, 2)\) and \(B(3, 4)\), we need to use the point-slope form or the two-point formula for the equation of a line.

#### Step 1: Calculate the slope (\(m\)) of the line.

The slope of the line through two points \((x_1, y_1)\) and \((x_2, y_2)\) is given by the formula:

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

Substitute the coordinates of points \(A(1, 2)\) and \(B(3, 4)\) into the formula:

$$
m = \frac{4 - 2}{3 - 1} = \frac{2}{2} = 1
$$

Thus, the slope of the line is \(m = 1\).

#### Step 2: Use the point-slope form to find the equation of the line.

The point-slope form of the equation of a line is given by:

$$
y - y_1 = m(x - x_1)
$$

We can use the point \(A(1, 2)\) and the slope \(m = 1\). Substituting these values into the point-slope form:

$$
y - 2 = 1(x - 1)
$$

Simplify the equation:

$$
y - 2 = x - 1
$$

Add 2 to both sides to solve for \(y\):

$$
y = x + 1
$$

#### Final Answer:

The equation of the line passing through points \(A(1, 2)\) and \(B(3, 4)\) is:

$$
y = x + 1
$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the points A and B
A = np.array([1, 2])
B = np.array([3, 4])

# Equation of the line y = x + 1
# Generate x values for the line
x_values = np.linspace(0, 4, 100)
y_values = x_values + 1

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the points A and B
ax.plot(A[0], A[1], 'ro', label='A(1, 2)')
ax.plot(B[0], B[1], 'bo', label='B(3, 4)')

# Plot the line y = x + 1
ax.plot(x_values, y_values, 'g-', label=r'$y = x + 1$')

# Set the limits of the graph
ax.set_xlim(0, 4)
ax.set_ylim(0, 5)

# Set the aspect of the plot to be equal
ax.set_aspect('equal')

# Add grid
ax.grid(True)

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

### 2. The line passes through point \(A(1, 2)\) and is parallel to the line \(y = 2x + 3\). Find the equation of the line.

To find the equation of the line that passes through point \(A(1, 2)\) and is parallel to the line \(y = 2x + 3\), we can use the fact that parallel lines have the same slope.

#### Step 1: Find the slope of the given line.

The given line is \(y = 2x + 3\). This is in the slope-intercept form of the equation of a line, \(y = mx + b\), where \(m\) is the slope and \(b\) is the y-intercept.

From the equation \(y = 2x + 3\), we can see that the slope of the line is:

$$
m = 2
$$

Since the new line is parallel to this line, it must have the same slope, \(m = 2\).

#### Step 2: Use the point-slope form to find the equation of the line.

The point-slope form of the equation of a line is given by:

$$
y - y_1 = m(x - x_1)
$$

We are given that the line passes through point \(A(1, 2)\), so we can use \(x_1 = 1\) and \(y_1 = 2\), and the slope \(m = 2\). Substituting these values into the point-slope form:

$$
y - 2 = 2(x - 1)
$$

Simplify the equation:

$$
y - 2 = 2x - 2
$$

Add 2 to both sides:

$$
y = 2x
$$

#### Final Answer:

The equation of the line passing through point \(A(1, 2)\) and parallel to the line \(y = 2x + 3\) is:

$$
y = 2x
$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the points and slope
A = np.array([1, 2])
slope = 2

# Equation of the given line: y = 2x + 3
# Generate x values for the line y = 2x + 3
x_values = np.linspace(0, 4, 100)
y_given_line = 2 * x_values + 3

# Equation of the new line: y = 2x
y_new_line = slope * x_values

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the given line y = 2x + 3
ax.plot(x_values, y_given_line, 'g-', label=r'$y = 2x + 3$')

# Plot the new line y = 2x (parallel line)
ax.plot(x_values, y_new_line, 'b--', label=r'$y = 2x$')

# Plot the point A(1, 2)
ax.plot(A[0], A[1], 'ro', label='A(1, 2)')

# Set the limits of the graph
ax.set_xlim(0, 4)
ax.set_ylim(0, 10)

# Set the aspect of the plot to be equal
ax.set_aspect('equal')

# Add grid
ax.grid(True)

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

### 3. To find the equation of the line that passes through point $A(1, 2)$ and is perpendicular to the line $y = 2x + 3$, follow these steps:

Step 1: Find the slope of the given line

The equation of the given line is $y = 2x + 3$, where the slope $m_1 = 2$.

Step 2: Use the perpendicular slope condition

For two lines to be perpendicular, the product of their slopes must be equal to $-1$. That is, if the slope of the given line is $m_1$, the slope of the perpendicular line $m_2$ satisfies:

$$
m_1 \cdot m_2 = -1
$$

Substitute $m_1 = 2$:

$$
2 \cdot m_2 = -1 \quad \Rightarrow \quad m_2 = -\frac{1}{2}
$$

Thus, the slope of the perpendicular line is $m_2 = -\frac{1}{2}$.

Step 3: Use the point-slope form of the line equation

The point-slope form of the equation of a line is:

$$
y - y_1 = m(x - x_1)
$$

where $(x_1, y_1)$ is a point on the line, and $m$ is the slope. Using point $A(1, 2)$ and slope $m_2 = -\frac{1}{2}$, the equation becomes:

$$
y - 2 = -\frac{1}{2}(x - 1)
$$

Step 4: Simplify the equation

Expand and simplify:

$$
y - 2 = -\frac{1}{2}x + \frac{1}{2}
$$

$$
y = -\frac{1}{2}x + \frac{1}{2} + 2
$$

$$
y = -\frac{1}{2}x + \frac{5}{2}
$$

Thus, the equation of the line is:

$$
y = -\frac{1}{2}x + \frac{5}{2}
$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the point A(1, 2)
A = np.array([1, 2])

# Equation of the given line: y = 2x + 3
# Generate x values for the line y = 2x + 3
x_values = np.linspace(0, 4, 100)
y_given_line = 2 * x_values + 3

# Equation of the perpendicular line: y = -1/2 * x + 5/2
y_perpendicular_line = -0.5 * x_values + 5 / 2

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the given line y = 2x + 3
ax.plot(x_values, y_given_line, 'g-', label=r'$y = 2x + 3$')

# Plot the perpendicular line y = -1/2 * x + 5/2
ax.plot(x_values, y_perpendicular_line, 'b--', label=r'$y = -\frac{1}{2}x + \frac{5}{2}$')

# Plot the point A(1, 2)
ax.plot(A[0], A[1], 'ro', label='A(1, 2)')

# Set the limits of the graph
ax.set_xlim(0, 4)
ax.set_ylim(0, 6)

# Set the aspect of the plot to be equal
ax.set_aspect('equal')

# Add grid
ax.grid(True)

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

### 4. To find the intersection point of the lines $y = 2x + 3$ and $y = 3x + 2$ and calculate the angle between them, follow these steps:


Step 1: Find the intersection point


To find the intersection point, set the equations of the lines equal to each other:

$$
2x + 3 = 3x + 2
$$

Now, solve for $x$:

$$
2x - 3x = 2 - 3
$$

$$
-x = -1
$$

$$
x = 1
$$

Now substitute $x = 1$ into one of the equations to find the $y$-coordinate. Using $y = 2x + 3$:

$$
y = 2(1) + 3 = 5
$$

Thus, the intersection point is $(1, 5)$.


Step 2: Find the slopes of the lines


The slope of the first line $y = 2x + 3$ is $m_1 = 2$.

The slope of the second line $y = 3x + 2$ is $m_2 = 3$.


Step 3: Calculate the angle between the two lines


The formula for the angle $\theta$ between two lines with slopes $m_1$ and $m_2$ is:

$$
\tan(\theta) = \left| \frac{m_1 - m_2}{1 + m_1 m_2} \right|
$$

Substitute $m_1 = 2$ and $m_2 = 3$:

$$
\tan(\theta) = \left| \frac{2 - 3}{1 + 2 \times 3} \right| = \left| \frac{-1}{7} \right| = \frac{1}{7}
$$

Now, find the angle $\theta$ by taking the inverse tangent:

$$
\theta = \tan^{-1} \left( \frac{1}{7} \right)
$$

Using a calculator:

$$
\theta \approx 8.13^\circ
$$

Thus, the angle between the two lines is approximately $8.13^\circ$.

\textbf{Final Answer:}

The intersection point of the lines is $(1, 5)$, and the angle between the lines is approximately $8.13^\circ$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Find the intersection point of y = 2x + 3 and y = 3x + 2
# Set the equations equal: 2x + 3 = 3x + 2
# Solve for x: x = 1
x_intersection = 1
y_intersection = 2 * x_intersection + 3  # Using y = 2x + 3

# The intersection point is (1, 5)
intersection_point = (x_intersection, y_intersection)

# Step 2: Find the slopes of the lines
m1 = 2  # Slope of y = 2x + 3
m2 = 3  # Slope of y = 3x + 2

# Step 3: Calculate the angle between the two lines
# Formula for the angle between two lines: tan(theta) = |(m1 - m2) / (1 + m1 * m2)|
tan_theta = abs((m1 - m2) / (1 + m1 * m2))

# Calculate theta (in radians) and convert to degrees
theta_rad = np.arctan(tan_theta)
theta_deg = np.degrees(theta_rad)

# Display the results
print(f"Intersection Point: {intersection_point}")
print(f"Angle between the lines: {theta_deg:.2f} degrees")

# Step 4: Plot the lines and the intersection point
x_values = np.linspace(-2, 4, 100)

# Define the equations of the lines
y_line1 = 2 * x_values + 3  # y = 2x + 3
y_line2 = 3 * x_values + 2  # y = 3x + 2

# Plot the lines
plt.plot(x_values, y_line1, label=r'$y = 2x + 3$', color='g')
plt.plot(x_values, y_line2, label=r'$y = 3x + 2$', color='b')

# Plot the intersection point
plt.plot(intersection_point[0], intersection_point[1], 'ro', label=f'Intersection (1, 5)')

# Set graph limits and labels
plt.xlim(-2, 4)
plt.ylim(-1, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

### 5. To find the equation of the line passing through the point $A(1, 2)$ and parallel to the vector $\mathbf{v} = [2, 3]$, we follow these steps:

Step 1: Understand the equation of the line

The general form of the equation of a line passing through a point $A(x_1, y_1)$ and parallel to a vector $\mathbf{v} = [v_1, v_2]$ is:

$$
\mathbf{r}(t) = (x_1, y_1) + t(v_1, v_2)
$$

where $t$ is a scalar parameter.

Step 2: Substitute the point $A(1, 2)$ and the vector $\mathbf{v} = [2, 3]$

We are given the point $A(1, 2)$ and the vector $\mathbf{v} = [2, 3]$, so the parametric equation of the line becomes:

$$
(x, y) = (1, 2) + t(2, 3)
$$

This gives the following system of equations for $x$ and $y$:

$$
x = 1 + 2t
$$

$$
y = 2 + 3t
$$

Step 3: Find the equation of the line in terms of $x$ and $y$

We now eliminate the parameter $t$ to express the equation of the line in standard form. From the equation for $x$:

$$
t = \frac{x - 1}{2}
$$

Substitute this into the equation for $y$:

$$
y = 2 + 3 \left( \frac{x - 1}{2} \right)
$$

Simplify the equation:

$$
y = 2 + \frac{3(x - 1)}{2}
$$

$$
y = 2 + \frac{3x - 3}{2}
$$

$$
y = 2 + \frac{3x}{2} - \frac{3}{2}
$$

$$
y = \frac{4}{2} + \frac{3x}{2} - \frac{3}{2}
$$

$$
y = \frac{3x + 1}{2}
$$

Final Answer

The equation of the line passing through the point $A(1, 2)$ and parallel to the vector $\mathbf{v} = [2, 3]$ is:

$$
y = \frac{3x + 1}{2}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Given point A(1, 2) and vector v = [2, 3]
A = np.array([1, 2])
v = np.array([2, 3])

# Parametric equations of the line
def parametric_line(t):
    return A + t * v

# Generate values for t
t_values = np.linspace(-5, 5, 400)

# Get x and y values for the line
line_points = np.array([parametric_line(t) for t in t_values])
x_values = line_points[:, 0]
y_values = line_points[:, 1]

# Step 2: Finding the standard equation y = (3x + 1)/2
# This is the equation we derived earlier
def equation_of_line(x):
    return (3 * x + 1) / 2

# Generate x values for plotting the equation
x_eq = np.linspace(-5, 5, 400)
y_eq = equation_of_line(x_eq)

# Plotting the line and point A
plt.plot(x_values, y_values, label=r'Line passing through A(1, 2) and parallel to v = [2, 3]', color='b')
plt.plot(x_eq, y_eq, label=r'Equation: $y = \frac{3x + 1}{2}$', linestyle='--', color='r')

# Mark point A(1, 2)
plt.scatter(1, 2, color='green', zorder=5, label="Point A(1, 2)")

# Set labels and plot limits
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

### 6. We have the line $y = 2x + 3$. Find an example of a line perpendicular and parallel to it.

Step 1: Determine the slope of the given line

The given line is in the form $y = mx + b$, where $m$ is the slope. For the line $y = 2x + 3$, the slope $m$ is:

$$
m = 2
$$

Step 2: Find the slope of the parallel line

The slope of a line parallel to the given line is the same as the slope of the given line. Therefore, the slope of the parallel line is also $2$.

So, the equation of a parallel line passing through any point $(x_1, y_1)$ will be:

$$
y - y_1 = 2(x - x_1)
$$

For example, if the parallel line passes through the point $(1, 4)$, the equation becomes:

$$
y - 4 = 2(x - 1)
$$

Simplifying:

$$
y = 2x + 2
$$

Thus, the equation of the parallel line is:

$$
y = 2x + 2
$$

Step 3: Find the slope of the perpendicular line

The slope of a line perpendicular to another line is the negative reciprocal of the original slope. Since the slope of the given line is $2$, the slope of the perpendicular line is:

$$
m_{\text{perpendicular}} = -\frac{1}{2}
$$

Step 4: Find the equation of the perpendicular line

The equation of a line with slope $-\frac{1}{2}$ passing through a point $(x_1, y_1)$ is:

$$
y - y_1 = -\frac{1}{2}(x - x_1)
$$

For example, if the perpendicular line passes through the point $(1, 4)$, the equation becomes:

$$
y - 4 = -\frac{1}{2}(x - 1)
$$

Simplifying:

$$
y - 4 = -\frac{1}{2}x + \frac{1}{2}
$$

$$
y = -\frac{1}{2}x + \frac{9}{2}
$$

Thus, the equation of the perpendicular line is:

$$
y = -\frac{1}{2}x + \frac{9}{2}
$$

Final Answer

- The equation of a line parallel to $y = 2x + 3$ is:
$$ y = 2x + 2 $$

- The equation of a line perpendicular to $y = 2x + 3$ is:
$$ y = -\frac{1}{2}x + \frac{9}{2} $$


```python
import numpy as np
import matplotlib.pyplot as plt

# Function for the original line y = 2x + 3
def original_line(x):
    return 2 * x + 3

# Function for the parallel line y = 2x + 2
def parallel_line(x):
    return 2 * x + 2

# Function for the perpendicular line y = -1/2x + 9/2
def perpendicular_line(x):
    return -0.5 * x + 9 / 2

# Generate x values for plotting
x_values = np.linspace(-5, 5, 400)

# Calculate y values for the lines
y_original = original_line(x_values)
y_parallel = parallel_line(x_values)
y_perpendicular = perpendicular_line(x_values)

# Plotting the lines
plt.plot(x_values, y_original, label=r'Original Line: $y = 2x + 3$', color='blue')
plt.plot(x_values, y_parallel, label=r'Parallel Line: $y = 2x + 2$', color='green')
plt.plot(x_values, y_perpendicular, label=r'Perpendicular Line: $y = -\frac{1}{2}x + \frac{9}{2}$', color='red')

# Mark the point (1, 4) where the parallel and perpendicular lines intersect
plt.scatter(1, 4, color='black', zorder=5, label="Point (1, 4)")

# Labels and grid
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)

# Add a legend
plt.legend()

# Show the plot
plt.show()

# Output the equations of the parallel and perpendicular lines
parallel_line_eq = "y = 2x + 2"
perpendicular_line_eq = "y = -1/2x + 9/2"

print(f"The equation of the parallel line is: {parallel_line_eq}")
print(f"The equation of the perpendicular line is: {perpendicular_line_eq}")
```

### 7. We have the line $y = 2x + 3$ and point $A(1, 2)$. Find the distance from point $A$ to the line. The formulaa;


$$
d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
$$

Step 1: Write the equation of the line in standard form

The given line is $y = 2x + 3$. To write it in the standard form $Ax + By + C = 0$, we rearrange the equation:

$$
y - 2x - 3 = 0
$$

So, we have:
$$
A = -2, \quad B = 1, \quad C = -3
$$

Step 2: Apply the formula for the distance

Now, we use the distance formula with the point $A(1, 2)$ and the coefficients from the equation of the line. Substituting $x_1 = 1$, $y_1 = 2$, $A = -2$, $B = 1$, and $C = -3$ into the formula:

$$
d = \frac{|(-2)(1) + (1)(2) - 3|}{\sqrt{(-2)^2 + 1^2}}
$$

Simplifying the numerator:

$$
d = \frac{|-2 + 2 - 3|}{\sqrt{4 + 1}}
$$

$$
d = \frac{|-3|}{\sqrt{5}} = \frac{3}{\sqrt{5}}
$$

Step 3: Simplify the result

To rationalize the denominator, we multiply both the numerator and the denominator by $\sqrt{5}$:

$$
d = \frac{3}{\sqrt{5}} \times \frac{\sqrt{5}}{\sqrt{5}} = \frac{3\sqrt{5}}{5}
$$

Final Answer

The distance from point $A(1, 2)$ to the line $y = 2x + 3$ is:

$$
d = \frac{3\sqrt{5}}{5}
$$


```python
import math

# Given line: y = 2x + 3
# Standard form: -2x + y - 3 = 0 (A = -2, B = 1, C = -3)

# Point A(1, 2)
x1, y1 = 1, 2

# Coefficients from the line equation
A = -2
B = 1
C = -3

# Applying the distance formula
distance = abs(A * x1 + B * y1 + C) / math.sqrt(A**2 + B**2)

# Output the result
print(f"The distance from point A(1, 2) to the line y = 2x + 3 is: {distance:.2f}")
```

### 8. The line intersects the coordinate axes at points $A(2, 0)$ and $B(0, 3)$. Find the equation of the line.

Step 1: Find the slope of the line

The slope of a line passing through two points $(x_1, y_1)$ and $(x_2, y_2)$ is given by the formula:

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

Substituting the coordinates of points $A(2, 0)$ and $B(0, 3)$:

$$
m = \frac{3 - 0}{0 - 2} = \frac{3}{-2} = -\frac{3}{2}
$$

So, the slope of the line is $m = -\frac{3}{2}$.

Step 2: Use the point-slope form to find the equation of the line

The point-slope form of the equation of a line is:

$$
y - y_1 = m(x - x_1)
$$

We can use point $A(2, 0)$ for this formula. Substituting $x_1 = 2$, $y_1 = 0$, and $m = -\frac{3}{2}$ into the equation:

$$
y - 0 = -\frac{3}{2}(x - 2)
$$

Simplifying:

$$
y = -\frac{3}{2}(x - 2)
$$

Step 3: Expand and simplify

Now, expand the equation:

$$
y = -\frac{3}{2}x + \frac{3}{2} \times 2
$$

$$
y = -\frac{3}{2}x + 3
$$

Final Answer:

The equation of the line is:

$$
y = -\frac{3}{2}x + 3
$$


```python
# Given points A(2, 0) and B(0, 3)

# Step 1: Calculate the slope (m)
x1, y1 = 2, 0  # Coordinates of point A
x2, y2 = 0, 3  # Coordinates of point B

# Calculate slope using the formula m = (y2 - y1) / (x2 - x1)
m = (y2 - y1) / (x2 - x1)

# Step 2: Use the point-slope form y - y1 = m(x - x1) to find the equation
# We will use point A(2, 0) for this formula
# y - 0 = m(x - 2)  =>  y = m(x - 2)
y_intercept = y1 - m * x1  # Calculate the y-intercept (b)

# Step 3: Display the slope and the equation of the line
print(f"The slope of the line is: {m}")
print(f"The equation of the line is: y = {m}x + {y_intercept}")
```

### 9. Calculate the angle between the line $y = x + 3$ and the $Ox$ axis.

Step 1: Find the slope of the line

The given equation of the line is in the slope-intercept form:

$$
y = mx + b
$$

where $m$ is the slope of the line. Comparing the given equation $y = x + 3$ with the general form, we can see that the slope $m$ of the line is:

$$
m = 1
$$

Step 2: Use the formula for the angle between a line and the $Ox$ axis

The angle $\theta$ between a line and the $Ox$ axis is given by the formula:

$$
\tan(\theta) = |m|
$$

where $m$ is the slope of the line. Substituting $m = 1$:

$$
\tan(\theta) = |1| = 1
$$

Step 3: Find the angle

To find the angle $\theta$, we take the inverse tangent (arctan) of 1:

$$
\theta = \tan^{-1}(1)
$$

Using the fact that $\tan^{-1}(1) = 45^\circ$, we have:

$$
\theta = 45^\circ
$$

Final Answer:

The angle between the line $y = x + 3$ and the $Ox$ axis is:

$$
\theta = 45^\circ
$$
