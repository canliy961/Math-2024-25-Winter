## 15. Equations of second-order curves (conic sections)

* Find the equation of a circle with center at point $A(1,2)$ and radius $r=3$.
* Find the equation of a parabola intersecting the $Ox$ axis at points $x=2$, $x=4$, and passing through point $y(3)=1$.
* Find the center of the ellipse with the equation $x^2 + 4y^2 - 4x - 16y + 16 = 0$.
* Find the slope ($m>0$) of the line $y=mx-5$ that is tangent to the circle with the equation $x^2 + y^2=1$.
* Find the intersection points of the hyperbola $x^2 - y^2 = 1$ with the ellipse's line $x^2 + 4y^2 = 6$.
* For the given hyperbola $x^2 - y^2 = 1$, find the distance between its branches.

### 1. To find the equation of a circle with center $A(1,2)$ and radius $r = 3$, we use the standard form of the equation of a circle:

$$(x - h)^2 + (y - k)^2 = r^2$$

where $(h, k)$ is the center and $r$ is the radius. Substituting $h = 1$, $k = 2$, and $r = 3$, we get:

$$(x - 1)^2 + (y - 2)^2 = 3^2$$

$$(x - 1)^2 + (y - 2)^2 = 9$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the center and radius of the circle
center = (1, 2)
radius = 3

# Create a set of angles to plot the circle
theta = np.linspace(0, 2 * np.pi, 100)

# Parametric equations for the circle
x = center[0] + radius * np.cos(theta)
y = center[1] + radius * np.sin(theta)

# Create the plot
plt.figure(figsize=(6, 6))
plt.plot(x, y, label="Circle with center (1, 2) and radius 3")

# Add the center point
plt.plot(center[0], center[1], 'ro', label="Center (1, 2)")

# Set the aspect ratio of the plot to be equal
plt.gca().set_aspect('equal', adjustable='box')

# Add labels and title
plt.title("Graph of the Circle (x - 1)^2 + (y - 2)^2 = 9")
plt.xlabel("x")
plt.ylabel("y")

# Add a grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

### 2. Given that the parabola intersects the $Ox$-axis at $x = 2$ and $x = 4$, and passes through the point $(3, 1)$, we can use the fact that the parabola is of the form:

$$
y = a(x - 2)(x - 4)
$$

Substituting the point $(3, 1)$ into the equation:

$$
1 = a(3 - 2)(3 - 4)
$$

$$
1 = a(1)(-1)
$$

$$
a = -1
$$

Thus, the equation of the parabola is:

$$
y = -(x - 2)(x - 4)
$$

Expanding the equation:

$$
y = -(x^2 - 6x + 8)
$$

$$
y = -x^2 + 6x - 8
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for the parabola
def parabola(x):
    return -x**2 + 6*x - 8

# Generate an array of x values
x_values = np.linspace(0, 6, 400)  # range from 0 to 6 for a clear view of the parabola

# Compute the corresponding y values
y_values = parabola(x_values)

# Create the plot
plt.figure(figsize=(6, 6))
plt.plot(x_values, y_values, label=r'$y = -x^2 + 6x - 8$', color='blue')

# Mark the roots (x-intercepts) at x = 2 and x = 4
plt.scatter([2, 4], [0, 0], color='red', zorder=5, label="Roots (2, 0) and (4, 0)")

# Mark the point (3, 1)
plt.scatter(3, 1, color='green', zorder=5, label="Point (3, 1)")

# Add labels and title
plt.title("Graph of the Parabola y = -x^2 + 6x - 8")
plt.xlabel("x")
plt.ylabel("y")

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.axhline(0, color='black',linewidth=1)  # x-axis
plt.axvline(0, color='black',linewidth=1)  # y-axis
plt.show()
```

### 3. The given equation of the ellipse is:

$$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$$

To find the center, we complete the square for both $x$ and $y$.

First, group the $x$-terms and $y$-terms:

$$
(x^2 - 4x) + 4(y^2 - 4y) = -16
$$

Complete the square for $x$ and $y$:

$$
x^2 - 4x = (x - 2)^2 - 4, \quad 4(y^2 - 4y) = 4((y - 2)^2 - 4) = 4(y - 2)^2 - 16
$$

Substitute back:

$$
(x - 2)^2 - 4 + 4(y - 2)^2 - 16 = -16
$$

Simplify:

$$
(x - 2)^2 + 4(y - 2)^2 = 4
$$

The center of the ellipse is $(2, 2)$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the ellipse
h, k = 2, 2  # Center of the ellipse
a = 2  # Semi-major axis (along the x-axis)
b = 1  # Semi-minor axis (along the y-axis)

# Generate an array of angles for parametric equations
theta = np.linspace(0, 2 * np.pi, 400)

# Parametric equations for the ellipse
x = h + a * np.cos(theta)
y = k + b * np.sin(theta)

# Create the plot
plt.figure(figsize=(6, 6))
plt.plot(x, y, label=r'$\frac{(x - 2)^2}{4} + \frac{(y - 2)^2}{1} = 1$', color='blue')

# Mark the center of the ellipse (2, 2)
plt.scatter(h, k, color='red', zorder=5, label="Center (2, 2)")

# Set the aspect ratio of the plot to be equal
plt.gca().set_aspect('equal', adjustable='box')

# Add labels and title
plt.title("Graph of the Ellipse")
plt.xlabel("x")
plt.ylabel("y")

# Add a grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

### 4. The equation of the circle is:

$$
x^2 + y^2 = 1
$$

The equation of the tangent line is:

$$
y = mx - 5
$$

For the line to be tangent to the circle, the distance from the center of the circle $(0, 0)$ to the line must be equal to the radius of the circle. The formula for the distance from a point $(x_1, y_1)$ to a line $Ax + By + C = 0$ is:

$$
d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
$$

For the line $y = mx - 5$, rewrite it in the form $mx - y - 5 = 0$ with $A = m$, $B = -1$, and $C = -5$. The distance from $(0, 0)$ to this line is:

$$
d = \frac{|m(0) - 1(0) - 5|}{\sqrt{m^2 + (-1)^2}} = \frac{5}{\sqrt{m^2 + 1}}
$$

Set this equal to the radius of the circle, which is 1:

$$
\frac{5}{\sqrt{m^2 + 1}} = 1
$$

Square both sides:

$$
25 = m^2 + 1
$$

$$
m^2 = 24
$$

$$
m = \pm \sqrt{24} = \pm 2\sqrt{6}
$$

Thus, the slope of the tangent line is:

$$
m = 2\sqrt{6}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the circle
theta = np.linspace(0, 2 * np.pi, 400)  # Parametric angle for the circle
x_circle = np.cos(theta)
y_circle = np.sin(theta)

# Define the slopes of the tangent lines
m1 = 2 * np.sqrt(6)
m2 = -2 * np.sqrt(6)

# Define the tangent lines: y = mx - 5
def tangent_line(m, x):
    return m * x - 5

# Create the x values for plotting the tangent lines
x_values = np.linspace(-3, 3, 400)

# Plot the circle
plt.figure(figsize=(6, 6))
plt.plot(x_circle, y_circle, label=r'Circle: $x^2 + y^2 = 1$', color='blue')

# Plot the tangent lines
plt.plot(x_values, tangent_line(m1, x_values), label=r'Tangent: $y = 2\sqrt{6}x - 5$', color='red', linestyle='--')
plt.plot(x_values, tangent_line(m2, x_values), label=r'Tangent: $y = -2\sqrt{6}x - 5$', color='green', linestyle='--')

# Mark the center of the circle (0, 0)
plt.scatter(0, 0, color='black', zorder=5, label="Center (0, 0)")

# Set the aspect ratio of the plot to be equal
plt.gca().set_aspect('equal', adjustable='box')

# Add labels and title
plt.title("Circle and Tangent Lines")
plt.xlabel("x")
plt.ylabel("y")

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

### 5. The equations of the hyperbola and ellipse are:

$$
x^2 - y^2 = 1, \quad x^2 + 4y^2 = 6
$$

To find the intersection points, subtract the first equation from the second:

$$
(x^2 + 4y^2) - (x^2 - y^2) = 6 - 1
$$

$$
5y^2 = 5
$$

$$
y^2 = 1
$$

$$
y = \pm 1
$$

Substitute $y = 1$ into the hyperbola equation:

$$
x^2 - 1^2 = 1
$$

$$
x^2 = 2 \quad \Rightarrow \quad x = \pm \sqrt{2}
$$

Substitute $y = -1$ into the hyperbola equation:

$$
x^2 - (-1)^2 = 1
$$

$$
x^2 = 2 \quad \Rightarrow \quad x = \pm \sqrt{2}
$$

Thus, the intersection points are:

$$
( \sqrt{2}, 1), (-\sqrt{2}, 1), (\sqrt{2}, -1), (-\sqrt{2}, -1)
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equations for the hyperbola and the ellipse
def hyperbola(x, y):
    return x**2 - y**2 - 1

def ellipse(x, y):
    return x**2 + 4*y**2 - 6

# Define the intersection points
intersection_points = [
    (np.sqrt(2), 1),
    (-np.sqrt(2), 1),
    (np.sqrt(2), -1),
    (-np.sqrt(2), -1)
]

# Generate values for plotting the hyperbola and ellipse
x_values = np.linspace(-3, 3, 400)
y_values_hyperbola_pos = np.sqrt(x_values**2 - 1)  # Positive branch of hyperbola
y_values_hyperbola_neg = -np.sqrt(x_values**2 - 1)  # Negative branch of hyperbola
y_values_ellipse_pos = np.sqrt((6 - x_values**2) / 4)  # Positive y branch of ellipse
y_values_ellipse_neg = -np.sqrt((6 - x_values**2) / 4)  # Negative y branch of ellipse

# Plot the hyperbola and ellipse
plt.figure(figsize=(6, 6))

# Plot the branches of the hyperbola
plt.plot(x_values, y_values_hyperbola_pos, label=r'Hyperbola: $x^2 - y^2 = 1$', color='blue')
plt.plot(x_values, y_values_hyperbola_neg, color='blue')

# Plot the branches of the ellipse
plt.plot(x_values, y_values_ellipse_pos, label=r'Ellipse: $x^2 + 4y^2 = 6$', color='red')
plt.plot(x_values, y_values_ellipse_neg, color='red')

# Mark the intersection points
for point in intersection_points:
    plt.scatter(point[0], point[1], color='green', zorder=5)
    plt.text(point[0] + 0.1, point[1], f'({point[0]:.2f}, {point[1]:.2f})', fontsize=9)

# Set the aspect ratio of the plot to be equal
plt.gca().set_aspect('equal', adjustable='box')

# Add labels and title
plt.title("Hyperbola and Ellipse with Intersection Points")
plt.xlabel("x")
plt.ylabel("y")

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

### 6. For the hyperbola

$$
x^2 - y^2 = 1
$$

the branches are asymptotic to the lines

$$
y = \pm x
$$

The distance between the branches is the distance between the two asymptotes at any point on the $x$-axis. At $y = 0$, the branches are at $x = \pm 1$. The distance between these points is:

$$
\text{Distance} = 1 - (-1) = 2
$$

Thus, the distance between the branches of the hyperbola is 2.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equation of the hyperbola and the asymptotes
def hyperbola_positive(x):
    return np.sqrt(x**2 - 1)

def hyperbola_negative(x):
    return -np.sqrt(x**2 - 1)

# Define the asymptotes y = x and y = -x
def asymptote_positive(x):
    return x

def asymptote_negative(x):
    return -x

# Generate x values for plotting
x_values = np.linspace(-3, 3, 400)
x_values_positive = x_values[x_values**2 >= 1]  # Ensure x^2 >= 1 for real values

# Generate y values for the hyperbola and asymptotes
y_values_hyperbola_pos = hyperbola_positive(x_values_positive)
y_values_hyperbola_neg = hyperbola_negative(x_values_positive)

# Plotting
plt.figure(figsize=(6, 6))

# Plot the branches of the hyperbola
plt.plot(x_values_positive, y_values_hyperbola_pos, label=r'Hyperbola: $x^2 - y^2 = 1$', color='blue')
plt.plot(x_values_positive, y_values_hyperbola_neg, color='blue')

# Plot the asymptotes
plt.plot(x_values, asymptote_positive(x_values), label=r'Asymptote: $y = x$', color='red', linestyle='--')
plt.plot(x_values, asymptote_negative(x_values), label=r'Asymptote: $y = -x$', color='green', linestyle='--')

# Mark the points where the branches intersect the x-axis
plt.scatter(1, 0, color='black', zorder=5)
plt.scatter(-1, 0, color='black', zorder=5)
plt.text(1.1, 0, '(1, 0)', fontsize=9)
plt.text(-1.1, 0, '(-1, 0)', fontsize=9)

# Add labels and title
plt.title("Hyperbola and Asymptotes with Distance Between Branches at y = 0")
plt.xlabel("x")
plt.ylabel("y")

# Set the aspect ratio of the plot to be equal
plt.gca().set_aspect('equal', adjustable='box')

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
